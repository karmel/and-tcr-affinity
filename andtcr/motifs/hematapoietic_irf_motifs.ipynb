```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set_context('poster')
```

```python
motifs = ['IRF2(IRF)/Erythroblas-IRF2-ChIP-Seq(GSE36985)/Homer',
          ]
names = ['IRF',]
```

```python
results = pd.io.parsers.read_csv('data/irf_motifs/hematapoietic_irf2.txt',
                sep='\t', 
                header=0)
print(results.columns)
results.head()
```

```python
cells = ['3T3L1', 'Th1', 'Th2', 'ESC', 'DN1', 'DN2a', 
         'DN2b', 'LSK', 'Stat1KO Th1', 'DP', 'ThioMac', 'DN3']
cells_ordered = ['ESC', 'LSK', 'DN1', 'DN2a', 'DN2b', 'DN3',
                 'DP', 'Th1', 'Th2', 'Stat1KO Th1', 'ThioMac', '3T3L1'][::-1]
results.index = cells
data = results.loc[cells_ordered]
data.head()
```

```python
target_f  = [float(s) for s in data['% of Target Sequences with Motif'].str.replace('%','')]
bg_f  = [float(s) for s in data['% of Background Sequences with Motif'].str.replace('%','')]
target_diff = [target_f[i] - bg_f[i] for i in range(len(target_f))]
target_f
```

```python

colorset = sns.color_palette('BuGn_d', len(results))
plt.figure()

width = .8
bg_base = [max(bg_f)]*len(data)
plt.barh(range(0,len(data)), bg_base, 
         width, color='#CCCCCC')
plt.barh(range(0,len(data)), target_diff, width, left=bg_base, color=colorset)

title = names[0] + ' motif frequency across hematopoietic cells'
plt.title(title, fontweight='bold')
plt.yticks([x + width/2 for x in range(0,len(data))], data.index)
plt.xlabel('Percent of H3K4me1/2 peaks with motif')
plt.ylim([-.2, len(data)])

plt.savefig('figures/{}.png'.format(title), bbox_inches='tight')
```
