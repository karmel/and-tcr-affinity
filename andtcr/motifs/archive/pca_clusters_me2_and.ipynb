```python
'''
Using the clusters generated by PCA,
create files for use with Homer motif finding.
'''
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set_context('poster')
```

```python
fields = ['chr_name', 'start', 'end', 'strand']
```

```python
for x in range(0,4):
    for section in ('bottom','top'):
        fname = 'pca_pc{}_{}_me2_and'.format(x + 1, section)
        data = pd.io.parsers.read_csv('../clustering/data/clusters/{}.txt'.format(fname), 
                        sep='\t', header=0)
        data = data.fillna(0)
        data['strand'] = 0
        data[fields].to_csv('output/{}.txt'.format(fname), 
                    sep='\t', 
                    header=False, index=True,
                    line_terminator='\r\n')
    
```

```python
'''
For parsing known results from the end of the PC that has the peptide-specific peaks:

mkdir specific_enriched_results
cp pca_pc1_top_me2_and_known/knownResults.txt specific_enriched_results/pca_pc1_me2_and_known.txt;
cp pca_pc2_bottom_me2_and_known/knownResults.txt specific_enriched_results/pca_pc2_me2_and_known.txt;
cp pca_pc3_top_me2_and_known/knownResults.txt specific_enriched_results/pca_pc3_me2_and_known.txt;
cp pca_pc4_bottom_me2_and_known/knownResults.txt specific_enriched_results/pca_pc4_me2_and_known.txt;
'''
```

```python
motifs = ['1-ACAGGAAGTG,BestGuess:ERG(ETS)/VCaP-ERG-ChIP-Seq(GSE14097)/Homer(0.984)',
          'RUNX1(Runt)/Jurkat-RUNX1-ChIP-Seq(GSE29180)/Homer',
          'IRF2(IRF)/Erythroblas-IRF2-ChIP-Seq(GSE36985)/Homer',
          'AP-1(bZIP)/ThioMac-PU.1-ChIP-Seq(GSE21512)/Homer',
          'Klf4(Zf)/mES-Klf4-ChIP-Seq(GSE11431)/Homer',
          'NFkB-p65-Rel(RHD)/LPS-exp(GSE23622)/Homer']
names = ['ETS', 'Runx', 'IRF', 'AP-1', 'KLF', 'NF-kB']

freqs = {}
bg_freqs = {}
logpvals = {}
for m in motifs:
    freqs[m] = []
    bg_freqs[m] = []
    logpvals[m] = []

pc_labels = ['PC{}'.format(x + 1) for x in range(0,4)]
for pc in range(0, len(pc_labels)):
    results = pd.io.parsers.read_csv('data/specific_enriched_results/pca_pc{}_me2_and_known.txt'.format(pc + 1), 
                    sep='\t', 
                    header=0, index_col=0)
    for m in motifs:
        target_f  = results.loc[m, '% of Target Sequences with Motif']
        bg_f  = results.loc[m, '% of Background Sequences with Motif']
        logp = results.loc[m, 'Log P-value']
        freqs[m].append(float(target_f.replace('%','')) - float(bg_f.replace('%','')))
        bg_freqs[m].append(float(bg_f.replace('%','')))
        logpvals[m].append(-logp)
        
colors = sns.color_palette('Blues', len(pc_labels))[::-1]
width = .8
for i, m in enumerate(motifs):
    plt.figure()
    
    plt.subplot(2,1,1)
    bg_base = [max(bg_freqs[m])]*len(pc_labels)
    plt.barh(range(0,len(pc_labels)), bg_base, 
             width, color='#CCCCCC')
    plt.barh(range(0,len(pc_labels)), freqs[m], width, color=colors, left=bg_base)
    plt.title(names[i] + ' Motif Frequency', fontweight='bold')
    plt.yticks([x + width/2 for x in range(0,len(pc_labels))], pc_labels)
    plt.xlabel('Percent of enhancers with motif')
    plt.ylim([-.2, len(pc_labels)])
    
    plt.subplot(2,1,2)
    plt.barh(range(0,len(pc_labels)), logpvals[m], width, color=colors)
    plt.title(names[i] + ' Motif Enrichment', fontweight='bold')
    plt.yticks([x + width/2 for x in range(0,len(pc_labels))], pc_labels)
    plt.xlabel('-Log(P-Value)')
    plt.ylim([-.2, len(pc_labels)])
    
    plt.tight_layout(h_pad=4)
    
    plt.savefig('figures/pca_{}_motif_frequency.png'.format(names[i]), bbox_inches='tight')
```
