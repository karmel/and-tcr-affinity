```python
'''
Using the clusters generated by kmeans clustering,
create files for use with Homer motif finding.
'''
```

```python
import pandas as pd
```

```python
k = 6
data = pd.io.parsers.read_csv('../clustering/data/clusters/kmeans_{}_me2_and.txt'.format(k), 
                                    sep='\t', 
                                    header=0)

data = data.fillna(0)
# Fill in strand col for ease of output file generation
data['strand'] = 0
```

```python
fields = ['chr_name', 'start', 'end', 'strand']
```

```python
for x in range(0,k):
    subdata = data[data['label'] == x]
    subdata[fields].to_csv('output/kmeans_{}_cluster_{}_me2_and.txt'.format(k, x+1), 
                    sep='\t', 
                    header=False, index=True,
                    line_terminator='\r\n')
```
