```python
'''
Generate files for motif finding with Homer.

Actual motif running done at the command line with Homer.

Note that Homer requires the \r line ending for all input files.
'''
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set_context('poster')
```

```python
data = pd.io.parsers.read_csv('../offstate/data/enhancers_me2_and_union.txt', 
                                    sep='\t', 
                                    header=0)

data = data.fillna(0)
# Fill in strand col for ease of output file generation
data['strand'] = 0
```

```python
peptides = ['no_peptide', '1um_k99a', '1um_pcc']
fields = ['chr_name', 'start', 'end', 'strand']
prefix = 'tag_count_'
```

```python
for pep in peptides:
    subdata = data[data[prefix + pep] > 0]
    subdata[fields].to_csv('output/enhancers_1um_all_{}.txt'.format(pep), 
                    sep='\t', 
                    header=False, index=True,
                    line_terminator='\r\n')
```

```python
for i, pep in enumerate(peptides):
    others = peptides[:i] + peptides[i+1:]
    other_cols = [prefix + o for o in others]
    subdata = data[data[other_cols].max(axis=1) == 0]
    subdata[fields].to_csv('output/enhancers_1um_only_{}.txt'.format(pep), 
                    sep='\t', 
                    header=False, index=True,
                    line_terminator='\r\n')
```

```python
'''
For parsing known results:

mkdir 1um_all_known_results 1um_only_known_results
for f in *all*known/knownResults.txt; 
do 
OUTPUT=`dirname ${f}`
cp $f 1um_all_known_results/${OUTPUT}.txt;
done;

for f in *only*known/knownResults.txt; 
do 
OUTPUT=`dirname ${f}`
cp $f 1um_only_known_results/${OUTPUT}.txt;
done;
'''
```

```python
motifs = ['1-ACAGGAAGTG,BestGuess:ERG(ETS)/VCaP-ERG-ChIP-Seq(GSE14097)/Homer(0.984)',
          'RUNX1(Runt)/Jurkat-RUNX1-ChIP-Seq(GSE29180)/Homer',
          'IRF2(IRF)/Erythroblas-IRF2-ChIP-Seq(GSE36985)/Homer',
          'AP-1(bZIP)/ThioMac-PU.1-ChIP-Seq(GSE21512)/Homer',
          'Klf4(Zf)/mES-Klf4-ChIP-Seq(GSE11431)/Homer',
          'NFkB-p65-Rel(RHD)/LPS-exp(GSE23622)/Homer']
names = ['ETS', 'Runx', 'IRF', 'AP-1', 'KLF', 'NF-kB']
labels = ['No Peptide', '1uM K99A', '1uM PCC']
```

```python


freqs = {}
bg_freqs = {}
logpvals = {}
for m in motifs:
    freqs[m] = []
    bg_freqs[m] = []
    logpvals[m] = []

for peptide in peptides:
    results = pd.io.parsers.read_csv('data/1um_all_known_results/enhancers_1um_all_{}_known.txt'.format(peptide), 
                    sep='\t', 
                    header=0, index_col=0)
    for m in motifs:
        target_f  = results.loc[m, '% of Target Sequences with Motif']
        bg_f  = results.loc[m, '% of Background Sequences with Motif']
        logp = results.loc[m, 'Log P-value']
        freqs[m].append(float(target_f.replace('%','')) - float(bg_f.replace('%','')))
        bg_freqs[m].append(float(bg_f.replace('%','')))
        logpvals[m].append(-logp)
        
colors = sns.color_palette('husl', len(peptides))[::-1]
width = .8
for i, m in enumerate(motifs):
    plt.figure()
    
    plt.subplot(2,1,1)
    bg_base = [max(bg_freqs[m])]*len(peptides)
    plt.barh(range(0,len(peptides)), bg_base, 
             width, color='#CCCCCC')
    plt.barh(range(0,len(peptides)), freqs[m], width, color=colors, left=bg_base)
    plt.title(names[i] + ' Motif Frequency', fontweight='bold')
    plt.yticks([x + width/2 for x in range(0,len(peptides))], labels)
    plt.xlabel('Percent of enhancers with motif')
    plt.ylim([-.2, len(peptides)])
    
    plt.subplot(2,1,2)
    plt.barh(range(0,len(peptides)), logpvals[m], width, color=colors)
    plt.title(names[i] + ' Motif Enrichment', fontweight='bold')
    plt.yticks([x + width/2 for x in range(0,len(peptides))], labels)
    plt.xlabel('-Log(P-Value)')
    plt.ylim([-.2, len(peptides)])
    
    plt.tight_layout(h_pad=4)
    
    plt.savefig('figures/1um_{}_motif_frequency.png'.format(names[i]), bbox_inches='tight')
```
