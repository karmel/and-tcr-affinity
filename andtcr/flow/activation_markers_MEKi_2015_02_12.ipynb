```python
'''
Plot output of FlowJo Geometric MFI analysis as a pretty bargraph.
'''
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set_context('poster')
```

```python
data = pd.io.parsers.read_csv('data/CD25_CD69_MFI_15-02-12.csv', 
                                    sep=',', 
                                    header=0, index_col=0)

# Remove Mean and SD rows
data = data[data['TUBE NAME'].isnull() == False]

# Remove PMA positive controls
data = data[data['TUBE NAME'].str[:3] != 'PMA']
```

```python
data.tail()
```

```python
# I want a particular ordering of columns.
col_order = [
        'No-Peptide', 'No-Peptide-0_5uM-MEKi', 
        '10uM-K99A', '10uM-K99A-0_5uM-MEKi',
        '100uM-K99A', '100uM-K99A-0_5uM-MEKi',
        '0_1uM-PCC', '0_1uM-PCC-0_5uM-MEKi',
        '10uM-PCC', '10uM-PCC-0_5uM-MEKi',
        ]
cols = ['No Peptide',
        '10uM K99A', 
        '100uM K99A', 
        '0.1uM PCC', 
        '10uM PCC', 
        ]
```

```python
data.index = data['TUBE NAME']
ordered = data.loc[col_order]
ordered.head()
```

```python
col_name = 'Lymphocytes/CD4+/CD69+ | Geometric Mean (Comp-FITC-A)'
ab = 'CD69'
```

```python
# Set up colors
colors = sns.color_palette("husl", 5)[::-1]

# Get sets of two
triples = zip(colors, colors)

# Flatten
triples = [item for sublist in triples for item in sublist]
```

```python
vals = ordered[col_name]

x_range = range(0, len(vals))
plt.figure()

bars = plt.bar([x + .2 for x in x_range], vals, color=list(triples))

# Add hatching on every other bar
for i, bar in enumerate(bars):
    if i%2 == 1:
        bar.set_hatch('//')
    else:
        bar.set_hatch(None)

# Custom legend
notx_patch = plt.Rectangle((0, 0), 1, 1, fc="white")
meki_patch = plt.Rectangle((0, 0), 1, 1, fc="white", hatch ='//')

plt.legend([notx_patch, meki_patch], ['Notx', '0.5uM MEKi'], loc='upper left')


plt.xlim([0, len(vals) + .2])
plt.ylabel('Mean Fluorescence Intensity for ' + ab)

# Complicated x-axis.
# First, the MEKi concentration
plt.xticks([x + 1.1 for x in x_range if not x%2], cols)

title = 'MFI of {}+ Population by Condition with MEK inhibitor'.format(ab)
plt.title(title)
plt.savefig('figures/{}.png'.format(title.replace(' ', '_')), bbox_inches='tight')
```
