```python
'''
Plot output of FlowJo Geometric MFI analysis as a pretty bargraph.
'''
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set_context('poster')
```

```python
data = pd.io.parsers.read_csv('data/CD25-CD69-Helios_MFI.csv', 
                                    sep=',', 
                                    header=0, index_col=0)

# Remove Mean and SD rows
data = data[data['TUBE NAME'].isnull() == False]

# Remove PMA positive controls
data = data[data['TUBE NAME'].str[:3] != 'PMA']
```

```python
data.tail()
```

```python
# Collapse NoPep tube names
data['TUBE NAME'] = data['TUBE NAME'].replace(to_replace='No-Peptide-\d', 
                                              value='No-Peptide',
                                              regex=True)
data['TUBE NAME'].unique().tolist()
```

```python
# I want a particular ordering of columns.
col_order = ['No-Peptide',
 'Y97K-1uM',
 'Y97K-10uM',
 'Y97K-100uM',
 'K99A-1uM',
 'K99A-10uM',
 'K99A-100uM',
 'K99E-1uM',
 'K99E-10uM',
 'K99E-100uM',
 'MCC-0_1uM',
 'MCC-10uM',
 'MCC-100uM',
 'PCC-0_0001uM',
 'PCC-0_001uM',
 'PCC-0_01uM',
 'PCC-0_1uM',
 'PCC-10uM',
 'PCC-100uM',
 ]

labels = [s.replace('-', ' ').replace('_', '.') for s in col_order]
peps = pd.Series(labels[:1] + [s.split(' ')[0] for s in labels[1:]]).unique()
concentrations = ['0uM'] + [s.split(' ')[1] for s in labels[1:]]
```

```python
ab = 'CD69'
col_name = 'Lymphocytes/CD4+/{}+ | Geometric Mean (Comp-FITC-A)'.format(ab)
```

```python
# Group by name
means = data.groupby(by='TUBE NAME')[col_name].mean()[col_order]
stddev = data.groupby(by='TUBE NAME')[col_name].std()[col_order]
```

```python
# Set up colors
colors = sns.color_palette("pastel", 6)[::-1]

# Get sets of six for the last five (leaving out no-pep)
colorset = zip(colors[1:], colors[1:], colors[1:])

# Flatten
colorset = colors[:1] + [item for sublist in colorset for item in sublist]

# Add an extra set for the last PCC samples
colorset += colorset[-1:]*3
```

```python
x_range = range(0, len(means))
plt.figure()

plt.bar([x + .2 for x in x_range], means, color=list(colorset))
plt.errorbar([x + .6 for x in x_range], means, linewidth=0,
        yerr=stddev, ecolor='black', elinewidth=.5, capsize=20)
plt.xlim([0, len(means) + .2])
plt.ylabel('Mean Fluorescence Intensity for ' + ab)

# Complicated x-axis.
plt.xticks([x + .6 for x in x_range], concentrations,
           rotation=90)

# Then, the peptide. Manually set because we have different counts per peptide...
label_loc = [.6, 2.7, 5.7, 8.7, 11.7, 16.2]
for i, loc in enumerate(label_loc):
    plt.text(loc, -1200, peps[i].replace(' ', '\n'), 
             horizontalalignment='center')

title = 'MFI of {}+ Population for All Peptides'.format(ab)
plt.title(title)
plt.savefig('figures/{}.png'.format(title.replace(' ', '_')), bbox_inches='tight')
```
