```python
'''
Plot output of FlowJo Geometric MFI analysis as a pretty bargraph.
'''
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.io.parsers.read_csv('data/pERK_MFI_2015_02_09.csv', 
                                    sep=',', 
                                    header=0, index_col=0)

# Remove Mean and SD rows
data = data[data['TUBE NAME'].isnull() == False]

# Remove PMA positive controls
data = data[data['TUBE NAME'].str[:3] != 'PMA']
```

```python
data.tail()
```

```python
# I want a particular ordering of columns.
col_order = [
        'No-Peptide-Notx', 'No-Peptide-MEKi-0_1uM', 'No-Peptide-MEKi-0_2uM',
        'No-Peptide-MEKi-0_5uM', 'No-Peptide-MEKi-1uM', 'No-Peptide-MEKi-10uM',
        'K99A-10uM-Notx', 'K99A-10uM-MEKi-0_1uM', 'K99A-10uM-MEKi-0_2uM', 
        'K99A-10uM-MEKi-0_5uM', 'K99A-10uM-MEKi-1uM', 'K99A-10uM-MEKi-10uM', 
        'K99A-100uM-Notx', 'K99A-100uM-MEKi-0_1uM', 'K99A-100uM-MEKi-0_2uM',
        'K99A-100uM-MEKi-0_5uM', 'K99A-100uM-MEKi-1uM', 'K99A-100uM-MEKi-10uM',
        'PCC-0_1uM-Notx', 'PCC-0_1uM-MEKi-0_1uM', 'PCC-0_1uM-MEKi-0_2uM', 
        'PCC-0_1uM-MEKi-0_5uM', 'PCC-0_1uM-MEKi-1uM', 'PCC-0_1uM-MEKi-10uM', 
        'PCC-10uM-Notx', 'PCC-10uM-MEKi-0_1uM', 'PCC-10uM-MEKi-0_2uM', 
        'PCC-10uM-MEKi-0_5uM', 'PCC-10uM-MEKi-1uM', 'PCC-10uM-MEKi-10uM',
        ]
cols = ['No Peptide',
        '10uM K99A', 
        '100uM K99A', 
        '0.1uM PCC', 
        '10uM PCC', 
        ]
meki = ['0uM', '0.1uM', '0.2uM', '0.5uM', '1uM','10uM']
```

```python
col_name = 'Lymphocytes/CD4+/pERK+ | Geometric Mean (Comp-FITC-A)'
ab = 'pERK'
```

```python
# Group by name
means = data.groupby(by='TUBE NAME')[col_name].mean()[col_order]
stddev = data.groupby(by='TUBE NAME')[col_name].std()[col_order]
```

```python
# Set up colors
# Get sets of six
color_groups = zip(colorset, colorset, colorset, colorset, colorset, colorset)

# Flatten
color_groups = [item for sublist in color_groups for item in sublist]
```

```python
labels = [tube.replace('-', ' ').replace('_', '.') for tube in col_order]
x_range = range(0, len(means))
plt.figure()

plt.bar([x + .2 for x in x_range], means, color=color_groups)
plt.errorbar([x + .6 for x in x_range], means, linewidth=0,
        yerr=stddev, ecolor='black', elinewidth=.5)
plt.xlim([0, len(means) + .2])
plt.ylabel('Mean Fluorescence Intensity for ' + ab)

# Complicated x-axis.
# First, the MEKi concentration
plt.xticks([x + .6 for x in x_range], meki*len(cols),
           rotation=90)
plt.text(-3, -400, 'MEKi\nconcentration', horizontalalignment='center')

# Then, the peptide
cond_per_pep = len(means)/len(cols)
label_loc = [x*cond_per_pep + cond_per_pep/2 for x in range(0,len(cols))]
for i, loc in enumerate(label_loc):
    plt.text(loc, -700, cols[i], horizontalalignment='center')

title = 'MFI of {}+ Population by Condition with MEK inhibitor'.format(ab)
plt.title(title, y=1.01)
plt.savefig('figures/{}'.format(title.replace(' ', '_')), bbox_inches='tight')
```
