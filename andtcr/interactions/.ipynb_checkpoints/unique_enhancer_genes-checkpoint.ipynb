```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set_context('poster')
```

```python
peptides = ['no_peptide', '10um_k99a', '100um_k99a', '0_1um_pcc', '10um_pcc']
labels = ['No Peptide', '10uM K99A', '100uM K99A', '0.1uM PCC', '10uM PCC']
peptide_cols = ['No-Peptide-24h', '10uM-K99A-24h', '100uM-K99A-24h', '0-1uM-PCC-24h', '10uM-PCC-24h',]
```

```python
max_dist = 50000
gene_sets = []
for peptide in peptides:
    data = pd.io.parsers.read_csv('data/gene_annotations/enhancers_only_{}_annotated.txt'.format(peptide), 
                                    sep='\t', 
                                    header=0)

    data = data.fillna(0)
    subdata = data[(data['Annotation'] != 'Intergenic') | 
                   (data['Distance to TSS'] < max_dist)]
    
    # Get counts of gene occurrances
    gene_sets.append(subdata['Gene Name'].value_counts())
```

```python
gene_df = pd.DataFrame(gene_sets, index=labels)
gene_df = gene_df.T.fillna(0)
gene_df.head()
```

```python
# Get Gene data
rpkm = pd.io.parsers.read_csv('../rna/data/mean/rpkm.txt',
                              sep='\t', header=0)
rpkm['gene_name'] = [s.split('|')[0] for s in rpkm['Annotation/Divergence']]
```

```python
# Set up colors
colorset = sns.color_palette("husl", 5)[::-1]
```

```python
min_rpkm = 10
for j, label in enumerate(labels):
    # For each condition, get the genes where there are the most
    # unique enhancers from that condition nearby.
    most = gene_df[gene_df[label] > gene_df.drop(label, axis=1).max(axis=1)]
    subdata = rpkm[rpkm['gene_name'].isin(most.index)]
    gene_names = subdata['gene_name']
    subdata = subdata[peptide_cols]
    subdata = subdata[subdata.min(axis=1) > min_rpkm]
    #subdata = subdata[subdata.max(axis=1) > 2*subdata.min(axis=1)]
    # Scale from 0 to 1.
    print(len(subdata))
    
    vals = subdata#.div(subdata.max(axis=1), axis=0).sort(peptide_cols[0])
    
    plt.figure()
    # Box plots of conditions
    bp = plt.boxplot([vals[p] for p in peptide_cols], patch_artist=True)

    # Colors.
    for i in range(0, len(colorset)):
        outline_col = '#666666'
        plt.setp(bp['boxes'][i], edgecolor=outline_col)
        plt.setp(bp['boxes'][i], linewidth=3)
        plt.setp(bp['boxes'][i], facecolor=colorset[i])
        plt.setp(bp['medians'][i], color=outline_col)
        plt.setp(bp['whiskers'][i*2], color=outline_col)
        plt.setp(bp['whiskers'][i*2+1], color=outline_col)
        plt.setp(bp['caps'][i*2], color=outline_col)
        plt.setp(bp['caps'][i*2+1], color=outline_col)


    plt.ylim([0,150])

    plt.ylabel('Normalized RPKM')
    plt.xticks([x + .6 for x in range(len(labels))], labels)

    title = 'Normalized RPKM at genes closest to {} unique enhancers'.format(labels[j])
    plt.title(title)

    plt.savefig('figures/{}.png'.format(title.replace(' ','_')),
                bbox_inches='tight')
```
