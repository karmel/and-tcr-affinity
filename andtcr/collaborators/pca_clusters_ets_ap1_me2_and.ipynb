```python
'''
Using the clusters generated by PCA,
determine if there is enrichment for Ets1 or AP-1 factors
that have been previously published.

CD4 Ets1
SRR561502
CD4/Treg input
SRR561496
Th17 Batf
SRR529957
Th17 JunD
SRR529961
Th17 cJun
SRR529960
Th17 JunD
SRR529959
Th17 Irf4
SRR529958
Th17 IgG
SRR529956
'''
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set_context('poster')
```

```python
peptides = ['no_peptide', '10um_k99a', '100um_k99a', '0_1um_pcc', '10um_pcc']
peaks = ['ets','irf4','batf','junb','cjun','jund']
suffix = '_tag_count'
min_tags = 0
width = .3
```

```python
for x in range(0,4):
    counts = {}
    for section in ('bottom','top'):
        fname = 'pca_pc{}_{}_me2_and'.format(x + 1, section)
        predata = pd.io.parsers.read_csv('../clustering/data/clusters/{}.txt'.format(fname), 
                        sep='\t', header=0, index_col=0)
        predata = predata.fillna(0)
              
        
        # We want the percentage of each cluster that has
        # each of the collaborating factors of interest.
        total = len(predata)
        counts[section] = []
        for peak in peaks:
            peak_count = sum(predata[peak + suffix] > min_tags)
            counts[section].append(peak_count/total*100)
    ax = plt.subplot(4,1, x+1)
    plt.bar(range(0, len(peaks)), 
                counts['top'],
                width,
                label='PC{} Top'.format(x + 1),
                color='#2b8cbe')
    plt.bar([i + width for i in range(0, len(peaks))], 
                counts['bottom'], 
                width,
                label='PC{} Bottom'.format(x + 1),
                color='#2ca25f')
    plt.legend(loc='upper left')
    plt.ylim([0,50])
    plt.yticks(range(10,60,10))
    plt.xticks([i + width for i in range(0, len(peaks))], peaks)
    if x != 3:
        plt.setp(ax.get_xticklabels(), visible=False)
    
```

```python
# We really want to compare the unique part of each PC, rather than top to bottom.
tag_counts = []
counts = []
for x in range(0,4):
    tag_counts.append({})
    counts.append({})
    for section in ('bottom','top'):
        fname = 'pca_pc{}_{}_me2_and'.format(x + 1, section)
        predata = pd.io.parsers.read_csv('../clustering/data/clusters/{}.txt'.format(fname), 
                        sep='\t', header=0, index_col=0)
        predata = predata.fillna(0)
        
        predata['ap1'] = predata[['batf_tag_count',
                    'cjun_tag_count',
                    'junb_tag_count',
                    'jund_tag_count',
                    ]].sum(axis=1)
        
        # We want the percentage of each cluster that has
        # each of the collaborating factors of interest.
        total = len(predata)
        peak_count = sum(predata['ap1'] > min_tags)
        counts[x][section] = peak_count/total*100
        tag_counts[x][section] = predata['ap1']

# Get the top of each PC except PC2, for which we want the bottom,
# as that is the unique portion.
unique_tag_counts = [tag_counts[0]['top'],
                     tag_counts[1]['bottom'],
                     tag_counts[2]['top'],
                     tag_counts[3]['bottom'],]
unique_counts = [counts[0]['top'],
                 counts[1]['bottom'],
                 counts[2]['top'],
                 counts[3]['bottom'],]

plt.figure()
plt.boxplot(unique_tag_counts)

plt.ylim([0,120])

plt.ylabel('Cumulative AP-1 Peak Tags')
plt.xticks([x+1 for x in range(0,len(counts))], 
           ['PC' + str(x+1) for x in range(0,len(counts))])

title = 'AP-1 enrichment in enhancers by principal component'
plt.title(title)
plt.savefig('figures/{}.png'.format(title.replace(' ','_')),
            bbox_inches='tight')

plt.figure()
colors = sns.color_palette('Blues_r', len(peptides))
width = .8
plt.bar(range(0, len(unique_counts)), 
                unique_counts,
                width, color=colors)

plt.ylabel('Percent of condition-specific 10% of PC with AP-1 binding')
plt.xticks([x+width/2 for x in range(0,len(counts))], 
           ['PC' + str(x+1) for x in range(0,len(counts))])


title = 'AP-1 frequency in enhancers by principal component'
plt.title(title)

plt.savefig('figures/{}.png'.format(title.replace(' ','_')),
            bbox_inches='tight')
```
