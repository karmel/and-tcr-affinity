```python
'''
For all enahncers from each peptide
determine if there is enrichment for Ets1 or AP-1 factors
that have been previously published.

CD4 Ets1
SRR561502
CD4/Treg input
SRR561496
Th17 Batf
SRR529957
Th17 JunD
SRR529961
Th17 cJun
SRR529960
Th17 JunD
SRR529959
Th17 Irf4
SRR529958
Th17 IgG
SRR529956
'''
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set_context('poster')
```

```python
peptides = ['no_peptide', '10um_k99a', '100um_k99a', '0_1um_pcc', '10um_pcc']
labels = ['No Peptide', '10uM K99A', '100uM K99A', '0.1uM PCC', '10uM PCC']
peaks = ['ets','irf4','batf','junb','cjun','jund']
prefix = 'tag_count_'
suffix = '_tag_count'
min_tags = 20
width = .15
```

```python
for me2_type in ('enhancers', 'promoters'):
    fname = '{}_me2_and_union'.format(me2_type)
    predata = pd.io.parsers.read_csv('../clustering/data/me2_and/{}.txt'.format(fname), 
                        sep='\t', header=0, index_col=0)
    predata = predata.fillna(0)
    
    plt.figure()
    colors = sns.color_palette('husl', len(peptides))[::-1]
    for i, peptide in enumerate(peptides):
        subdata = predata[(predata[prefix + peptide] > min_tags)]
        
        # Unique enhancers
        others = [prefix + p for p in peptides[:i] + peptides[i+1:]]
        subdata = subdata[subdata[others].max(axis=1) <= min_tags]
    
        total = len(subdata)
        counts = []
        
        # Get percent of enhancers with each peak
        for peak in peaks:
            peak_count = sum(subdata[peak + suffix] > min_tags)
            counts.append(peak_count/total*100)

        plt.bar([x + i*width for x in range(0, len(counts))], 
                counts,
                width,
                label=peptide,
                color=colors[i])
    plt.legend(loc='upper left')
    #plt.ylim([0,50])
    #plt.yticks(range(10,60,10))
    plt.xticks([i + width for i in range(0, len(peaks))], 
               [p.title() for p in peaks])
    
```

```python
fname = 'enhancers_me2_and_union'
predata = pd.io.parsers.read_csv('../clustering/data/me2_and/{}.txt'.format(fname), 
                    sep='\t', header=0)
predata = predata.fillna(0)

predata['ap1'] = predata[['batf_tag_count',
                    'cjun_tag_count',
                    'junb_tag_count',
                    'jund_tag_count',
                    ]].sum(axis=1)

plt.figure()
colors = sns.color_palette('husl', len(peptides))[::-1]

tag_counts = []
ap1_presence = []
for i, peptide in enumerate(peptides):
    subdata = predata[(predata[prefix + peptide] > min_tags)]

    # Unique enhancers
    others = [prefix + p for p in peptides[:i] + peptides[i+1:]]
    subdata = subdata[subdata[others].max(axis=1) <= min_tags]
    tag_counts.append(subdata['ap1'])

    total = len(subdata)
    with_ap1 = sum(subdata['ap1'] > min_tags)
    ap1_presence.append(with_ap1/total*100)

plt.figure()
plt.boxplot(tag_counts)

plt.ylim([0,120])

plt.ylabel('Cumulative AP-1 Peak Tags')
plt.xticks(range(1, len(labels) + 1), labels)

title = 'AP-1 enrichment in enhancers by condition'
plt.title(title)
plt.savefig('figures/{}.png'.format(title.replace(' ','_')),
            bbox_inches='tight')

plt.figure()
colors = sns.color_palette('husl', len(peptides))[::-1]
width = .8
plt.bar(range(0, len(ap1_presence)), 
                ap1_presence,
                width, color=colors)

plt.ylabel('Percent of condition-specific enhancers with AP-1 binding')
plt.xticks([x + width/2 for x in range(0, len(peptides))], labels)

title = 'AP-1 frequency in enhancers by condition'
plt.title(title)

plt.savefig('figures/{}.png'.format(title.replace(' ','_')),
            bbox_inches='tight')
```
