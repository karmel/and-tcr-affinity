```python
'''
What are the genes around appearing/disappearing superenhancers?
'''
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
```

```python
filename = 'data/no_peptide_1um_pcc_superenhancers_merged.txt'
peaks = pd.io.parsers.read_csv(filename, 
                                    sep='\t', 
                                    header=0, index_col=0)

filename = 'data/no_peptide_1um_pcc_superenhancers_annotated_merged.txt'
annot = pd.io.parsers.read_csv(filename, 
                                    sep='\t', 
                                    header=0, index_col=0)
```

```python
data = pd.merge(peaks, annot, left_index=True, right_index=True)
data = data.fillna(0)
```

```python
print(data.columns)
data.describe()
```

```python
pcc = data[data['AND-CD4TCell-H3K27Ac-No-Peptide-24h-MNase-Combined_superenhancers.txt'] == 0]
pcc.describe()
```

```python
nopep = data[data['AND-CD4TCell-H3K27Ac-1uM-PCC-24h-MNase-Combined_superenhancers.txt'] == 0]
nopep.describe()
```

```python
shared = data[(data['AND-CD4TCell-H3K27Ac-No-Peptide-24h-MNase-Combined_superenhancers.txt'] != 0) 
              & (data['AND-CD4TCell-H3K27Ac-1uM-PCC-24h-MNase-Combined_superenhancers.txt'] != 0)]
shared.describe()
```

```python
# Gene lists for go.
pcc.to_csv('output/1um_pcc_only_superenhancers.txt', sep='\t',
             header=True, index=True)
nopep.to_csv('output/no_peptide_only_superenhancers.txt', sep='\t',
             header=True, index=True)
shared.to_csv('output/shared_superenhancers.txt', sep='\t',
             header=True, index=True)
```

```python
# Not in shared but in PCC only?
pcc['Gene Name'][pcc['Gene Name'].isin(shared['Gene Name'])]
```

```python
# Enrichment for activation signature genes?
# Get genes from PC1
top_genes = pd.io.parsers.read_csv('../rna/unsupervised/output/clusters/pca_pc1_top_rna_and.txt', 
                                    sep='\t', 
                                    header=0, index_col=0)
bottom_genes = pd.io.parsers.read_csv('../rna/unsupervised/output/clusters/pca_pc1_bottom_rna_and.txt', 
                                    sep='\t', 
                                    header=0, index_col=0)
```

```python
# Get peak files so that we have tag counts
nopep_peaks = pd.io.parsers.read_csv('data/AND-CD4TCell-H3K27Ac-No-Peptide-24h-MNase-Combined_superenhancers.txt', 
                                    sep='\t', 
                                    header=None, index_col=0, comment='#')
columns = 'chr	start	end	strand	Normalized Tag Count	superEnhancer slope	findPeaks Score	Total Tags'.split('\t')
nopep_peaks.columns = columns

pcc_peaks = pd.io.parsers.read_csv('data/AND-CD4TCell-H3K27Ac-1uM-PCC-24h-MNase-Combined_superenhancers.txt', 
                                    sep='\t', 
                                    header=None, index_col=0, comment='#')
pcc_peaks.columns = columns
```

```python
# Merge in the tag counts.
shared_peaks = pd.merge(shared, pcc_peaks, 
                        left_on='AND-CD4TCell-H3K27Ac-1uM-PCC-24h-MNase-Combined_superenhancers.txt',
                        right_index=True,
                        suffixes=('_shared', '_pcc'))
shared_peaks = pd.merge(shared_peaks, nopep_peaks, 
                        left_on='AND-CD4TCell-H3K27Ac-No-Peptide-24h-MNase-Combined_superenhancers.txt',
                        right_index=True,
                        suffixes=('_pcc', '_nopep'))
shared_peaks.columns
```

```python
# Now calculate ratio of tag counts
shared_peaks['log2'] = np.log2(shared_peaks['Normalized Tag Count_pcc']/
                               shared_peaks['Normalized Tag Count_nopep'])
shared_peaks['log2'].head()
```

```python
# Split into groups
top_enh = shared_peaks[shared_peaks['Gene Name'].isin(top_genes['gene_name'])]
rest = shared_peaks[shared_peaks['Gene Name'].isin(top_genes['gene_name']) == False]
```

```python
# Set up colors
labels = ['Not activation signature', 'Activation signature']
```

```python
# Box plots of conditions
bp = plt.boxplot([rest['log2'], top_enh['log2']], sym='')

plt.ylim([-1,1.5])

plt.ylabel('Log2(1uM PCC tag count/No Peptide tag count)')
plt.xticks(range(1,len(labels)+1), labels)

title = 'Ratio of H3K27Ac tag counts at shared super-enhancers'
plt.title(title, y=1.01)

plt.savefig('figures/{}'.format(title.replace(' ','_')),
            bbox_inches='tight')
```

```python
# Pairwise statistical significance?
from scipy.stats import ttest_ind

print(ttest_ind(rest['log2'], top_enh['log2']))
```

```python
# Do denovo show enrichment for top genes?
top_ct = sum(pcc['Gene Name'].isin(top_genes['gene_name']))
top_ct
```

```python
act_only_not_top = len(pcc) - top_ct
existing_top = len(top_genes) - top_ct
existing_not_top = len(data['Gene Name'].unique())
stats.chi2_contingency([[top_ct, existing_top], 
                             [act_only_not_top, existing_not_top]])
```

```python
# What about shared?
top_ct = len(top_enh)
act_only_not_top = len(shared) - top_ct
existing_top = len(top_genes) - top_ct
existing_not_top = len(data['Gene Name'].unique())
stats.chi2_contingency([[top_ct, existing_top], 
                             [act_only_not_top, existing_not_top]])
```

```python
# What about shared?
bottom_ct = sum(shared['Gene Name'].isin(bottom_genes['gene_name']))
act_only_not_bottom = len(shared) - bottom_ct
existing_bottom = len(bottom_genes) - bottom_ct
existing_not_bottom = len(data['Gene Name'].unique())
stats.chi2_contingency([[bottom_ct, existing_bottom], 
                             [act_only_not_bottom, existing_not_bottom]])
```
