'''
We want to plot the relative expression levels of AP-1 factors,
comparing WT and KO across the various conditions.

We will do this as line plots first, but there is likely a 
more elegant way to present this data given that there are three
binary variables: wt/ko, 0d/7d, klrghi/klrglo.
'''

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set_context('poster')
```

```python
filename = '../data/rpkm.txt'
data = pd.io.parsers.read_csv(filename, 
                                    sep='\t', 
                                    header=0)
data.columns
```

```python
# Clean up column names for easier parsing.
new_index = []
for col in data.columns:
    cleaned = col.replace('-KAA-14-10-31 reads','').replace(' FPKM','')
    cleaned = cleaned.replace('CD8TCell-','').replace('RNA-','')
    new_index.append(cleaned)
new_index[0] = 'id'
data.columns = new_index
data.columns
```

```python
# Extract the genes we care about.
factor_pairs = ((('Fos'),('NM_010234')),
           (('Fosb'),('NM_008036')),
           (('Fosl1'),('NM_010235')),
           (('Fosl2'),('NM_008037')),
           (('Jun'),('NM_010591')),
           (('Junb'),('NM_008416')),
           (('Jund'),('NM_001286944')),
           (('Jdp2'),('NM_001205052')),
           (('Atf1'),('NM_007497')),
           (('Atf2'),('NM_001025093')),
           (('Atf3'),('NM_007498')),
           (('Atf4'),('NM_001287180')),
           (('Atf5'),('NM_030693')),
           (('Atf6'),('NM_001081304')),
           (('Atf7'),('NM_146065')),
           (('Batf'),('NM_016767')),
           (('Batf2'),('NM_028967')),
           (('Batf3'),('NM_030060')),
           )
zipped = list(zip(*factor_pairs))
factors = pd.DataFrame({'gene_name':zipped[0],
                        'id':zipped[1],})
factors
```

```python
subdata = pd.merge(factors, data, on='id', how='inner')
subdata
```

```python
# Isolate and order columns we care about
cols = ['LCMV-0d', 
        'KLRGhi-LCMV-7d', 
        'KLRGlo-LCMV-7d', 
        'KLRGhi-LCMV-12d', 
        'KLRGlo-LCMV-12d', 
        'LCMV-35d-rep2', 
        'Foxo1-KO-LCMV-0d',
        'Foxo1-KO-KLRGhi-LCMV-7d', 
        'Foxo1-KO-KLRGlo-LCMV-7d', 
        'Foxo1-KO-KLRGhi-LCMV-12d', 
        ]
gene_data = subdata[cols]
gene_data.index = subdata['gene_name']
```

```python
# First, plot all conditions for each gene
sns.set_palette('husl', 14)
gene_data.transpose().plot()
plt.xticks(rotation=90)
title = 'AP-1 Factor RPKM for All Conditions'
plt.title(title)
plt.ylabel('RPKM')
plt.savefig('figures/' + title.replace(' ', '_') + '.png',
            bbox_inches='tight')
```

```python
of_interest = ['Fos','Jun','Junb']
genes_sub = gene_data.loc[of_interest]
sns.set_palette('husl', 3)
genes_sub.transpose().plot()
plt.xticks(rotation=90)
title = 'AP-1 Factors of Interest for All Conditions'
plt.title(title)
plt.ylabel('RPKM')
plt.savefig('figures/' + title.replace(' ', '_') + '.png',
            bbox_inches='tight')
```

```python
# Now split to overlay WT and KO
wt = genes_sub[['LCMV-0d', 'KLRGhi-LCMV-7d', 'KLRGlo-LCMV-7d', 
        'KLRGhi-LCMV-12d', 'KLRGlo-LCMV-12d', 'LCMV-35d-rep2']]
ko = genes_sub[['Foxo1-KO-LCMV-0d', 'Foxo1-KO-KLRGhi-LCMV-7d', 
        'Foxo1-KO-KLRGlo-LCMV-7d', 'Foxo1-KO-KLRGhi-LCMV-12d']]
ko.index = ['Foxo1-KO-' + g for g in ko.index]

# Make colors
palette = sns.color_palette('YlOrBr_d', 3) + sns.color_palette('BuGn_d', 3)
sns.set_palette(palette)
ax = ko.transpose().plot()
wt.transpose().plot(ax=ax)
plt.xticks(rotation=90)
title = 'AP-1 Factors of Interest: WT vs. KO'
plt.title(title)
plt.ylabel('RPKM')
plt.savefig('figures/' + title.replace(' ', '_') + '.png',
            bbox_inches='tight')
```

```python
sns.set_palette('hls',2)

plt.figure()
for i, gene in enumerate(wt.index):
    plt.subplot(len(wt.index)*100 + 11 + i)
    ko.loc['Foxo1-KO-' + gene].plot()
    wt.loc[gene].plot()
    plt.title(gene + ': WT vs. KO')
    plt.ylabel('RPKM')
    plt.legend()
    
    if i + 1 == len(wt.index):
        plt.xticks(rotation=90)
    else:
        plt.xticks([])

plt.savefig('figures/AP-1_Factors_of_Interest_Comparison_Grid.png',
            bbox_inches='tight')
```
