```python
'''
We have a set of ~300 genes with names and scores.

Given a microarray .txt file, normalize, extract the genes of interest,
and determine score.
'''
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import subprocess
```

```python
sns.set_context('poster')
```

```python
gse = 'GSE37554'
fname = 'data/{}_series_matrix.txt'.format(gse)

# Comment lines begin with a variety of characters. Figure out where to start manually.
start_at = 0
for i, line in enumerate(open(fname).readlines()):
    if line.startswith('!Sample_title'):
        headers = line.split('\t')[1:]
    elif line.startswith('!Series_platform_id'):
        platform = line.split('\t')[1].strip().strip('"')
    elif line[0] not in ('!', '#', '^', '\n'):
        start_at = i
        break
data = pd.io.parsers.read_csv(fname, 
                sep='\t', header=start_at-1, index_col=0)
data = data.fillna(0)
headers = [h.strip().strip('"').replace(' ','_') for h in headers]
print(platform)
print(headers)
data.head()
```

```python
headers_ordered = ['5388876092_A', '5388876092_B', 
                   '5388876092_C', '5388876092_D', 
                   '5388876092_E', '5388876092_F',
                   '5388876097_G', '5388876097_H',
                   '5388876092_G', '5388876092_H', 
                   '5388876097_C', '5388876097_D',                   
                   '5388876097_A', '5388876097_B', 
                   '5388876097_E', '5388876097_F', 
                   ]
labels = ['d0', 'Th0 3h', 'iTreg 3h', 'Th0 3h\n+ TGF',
          'Th0 d1', 'iTreg d1', 'Th1 d1', 'Th0 d1\n+ TGF',]
```

```python
# Rename our columns
data.columns = headers
data.head()
```

```python
# Get gene names
fname = 'data/{}.txt'.format(platform)
start_at = 0
for i, line in enumerate(open(fname).readlines()):
    if line[0] not in ('!', '#', '^', '\s', '\n'):
        start_at = i
        break
probes = pd.io.parsers.read_csv(fname, 
            sep='\t', header=start_at, index_col=0)
probes.head()
```

```python
probes['Gene symbol'] = probes['Symbol']
probes.head()
```

```python
merged = pd.merge(data, probes, left_index=True, right_index=True)
merged.head()
```

```python
subdata = merged[['Gene symbol'] + headers_ordered]
filename = 'data/{}_gene_values.txt'.format(gse)
subdata.to_csv(filename,
               header=True, index=False, sep='\t')
scores = subprocess.check_output('python ../calculate_activation_signature_score.py -f ' +
                                '{}'.format(filename), shell=True)
print(scores)
```

```python
import sys
sys.path.append('../')
from calculate_activation_signature_score import ActivationOptionParser

parser = ActivationOptionParser()
filename = 'data/{}_gene_values.txt'.format(gse)
scores = parser.scores_from_file(filename)
```

```python
# Set up colors
colors = sns.color_palette("GnBu_d", len(scores)//2)

# Get sets of two
triples = zip(colors, colors)

# Flatten
triples = [item for sublist in triples for item in sublist]
```

```python
x_range = range(0, len(scores))
plt.figure()

bars = plt.bar([x + .2 for x in x_range], scores, color=list(triples))

# Add hatching on every other bar
for i, bar in enumerate(bars):
    if i%2 == 1:
        bar.set_hatch('//')
    else:
        bar.set_hatch(None)

# Custom legend
notx_patch = plt.Rectangle((0, 0), 1, 1, fc="white")
meki_patch = plt.Rectangle((0, 0), 1, 1, fc="white", hatch ='//')

plt.legend([notx_patch, meki_patch], ['Notx', 'Erk2 KO'], loc='upper left')

# Label scores above bars
for i, score in enumerate(scores):
    if score < 0:
        align = 'top'
        height = score - 15
    else: 
        align = 'bottom'
        height = score + 1
    plt.text(i + .6, height, int(score),
                ha='center', va=align, fontsize=14)

plt.ylim([plt.ylim()[0], plt.ylim()[1] + 100])
plt.xlim([0, len(scores) + .2])
plt.ylabel('Activation Signature Score')

plt.xticks([x + 1.1 for x in x_range if x%2], ['']*len(labels))
for i, label in enumerate(labels):
    plt.text(i*2 + 1.1, plt.ylim()[0]*1.01, label,
                ha='center', va='top', fontsize=14)

title = 'Activation Signature Scores: {}'.format(gse)
plt.title(title)
plt.savefig('figures/{}.png'.format(title.replace(' ', '_')), bbox_inches='tight')
```
