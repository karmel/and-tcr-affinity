```python
'''
We have a set of ~300 genes with names and scores.

Given a microarray .txt file, normalize, extract the genes of interest,
and determine score.
'''
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import subprocess
import numpy as np
```

```python
sns.set_context('poster')
```

```python
gse = 'GSE42276'
fname = 'data/{}_series_matrix.txt'.format(gse)

# Comment lines begin with a variety of characters. Figure out where to start manually.
start_at = 0
for i, line in enumerate(open(fname).readlines()):
    if line.startswith('!Sample_title'):
        headers = line.split('\t')[1:]
    elif line.startswith('!Series_platform_id'):
        platform = line.split('\t')[1].strip().strip('"')
    elif line[0] not in ('!', '#', '^', '\n'):
        start_at = i
        break
data = pd.io.parsers.read_csv(fname, 
                sep='\t', header=start_at-1, index_col=0)
data = data.fillna(0)
headers = [h.strip().strip('"').replace(' ','_').replace(',','') for h in headers]
print(platform)
print(headers)
data.head()
```

```python
headers_ordered = headers
```

```python
# Rename our columns
data.columns = headers
data.head()
```

```python
# Get gene names
fname = 'data/{}.txt'.format(platform)
start_at = 0
for i, line in enumerate(open(fname).readlines()):
    if line[0] not in ('!', '#', '^', '\s', '\n'):
        start_at = i
        break
probes = pd.io.parsers.read_csv(fname, 
            sep='\t', header=start_at, index_col=0)
probes.head()
```

```python
def get_symbol(row):
    try:
        return str(row['gene_assignment']).split('//')[1].strip()
    except IndexError:
        return ''

probes['Gene symbol'] = probes.apply(get_symbol, axis=1)
probes.head()
```

```python
merged = pd.merge(data, probes, left_index=True, right_index=True)
merged.head()
```

```python
subdata = merged[['Gene symbol'] + headers_ordered]
filename = 'data/{}_gene_values.txt'.format(gse)
subdata.to_csv(filename,
               header=True, index=False, sep='\t')
scores = subprocess.check_output('python ../calculate_activation_signature_score.py -f ' +
                                '{}'.format(filename), shell=True)
print(scores)
```

```python
import sys
sys.path.append('../')
from calculate_activation_signature_score import ActivationOptionParser

parser = ActivationOptionParser()
filename = 'data/{}_gene_values.txt'.format(gse)
scores = parser.scores_from_file(filename)
```

```python
# Irregular numbers of samples-- calc means and std.
groups = [scores[:3]] # Unstim
groups += [scores[15:19]] + [scores[19:23]] # CD3 only
groups += [scores[9:12]] + [scores[12:15]] # CD28
groups += [scores[3:6]] + [scores[6:9]]
groups += [scores[23:25]] + [scores[25:27]]
groups += [scores[27:30]] + [scores[30:33]]
groups += [scores[33:35]] + [scores[35:37]]
groups += [scores[37:40]] + [scores[40:43]]

means = [np.mean(score_group) for score_group in groups]
stddev = [np.std(score_group) for score_group in groups]

labels = ['Unstim', 'CD3 only', '+ CD28',
          '+ BTLA', '+ CD80',
          '+ CTLA4', '+ ICOS', 
          '+ PD1']

groups
```

```python
# Set up colors
colorset = sns.color_palette("GnBu_d", len(means)//2)

# Get sets of two
color_groups = zip(colorset[1:], colorset[1:])

# Flatten
color_groups = [colorset[0]] + [item for sublist in color_groups for item in sublist]
```

```python
x_range = range(0, len(means))
plt.figure()

bars = plt.bar([x + .2 for x in x_range], means, color=list(color_groups))
plt.errorbar([x + .6 for x in x_range], means, linewidth=0,
        yerr=stddev, ecolor='black', elinewidth=.5, capsize=20)

# Add hatching on every other bar
for i, bar in enumerate(bars):
    if i == 0:
        continue # Unstim
    elif i%2 == 1:
        bar.set_hatch(None)
    else:
        bar.set_hatch('//')

# Custom legend
clear_patch = plt.Rectangle((0, 0), 1, 1, fc="white")
hatch_patch = plt.Rectangle((0, 0), 1, 1, fc="white", hatch ='//')

plt.legend([clear_patch, hatch_patch], 
           ['Early timepoint', 'Late timepoint'], loc='lower left')

# Label scores above bars
for i, score in enumerate(means):
    if score < 0:
        align = 'top'
        height = score - .1
    else: 
        align = 'bottom'
        height = score + .1
    plt.text(i + .6, height, '%.2f' % score,
                ha='center', va=align, fontsize=14)

plt.ylim([-1.1, 1.1])
plt.xlim([0, len(means) + .2])
plt.ylabel('Activation Signature Score')

plt.xticks([1.1] + [x + 1.1 for x in x_range[1:] if not x%2], ['']*len(labels))
for i, label in enumerate(labels):
    if i == 0:
        x_loc = .6
    else:
        x_loc = i*2 + .1
    plt.text(x_loc, plt.ylim()[0]*1.01, label,
                ha='center', va='top', fontsize=14)


title = 'Activation Signature Scores: {}'.format(gse)
plt.title(title)
plt.savefig('figures/{}.png'.format(title.replace(' ', '_')), bbox_inches='tight')
```
