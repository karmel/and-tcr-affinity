```python
'''
We have a set of ~300 genes with names and scores.

Given a microarray .soft file, normalize, extract the genes of interest,
and determine score.
'''
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import subprocess
```

```python
sns.set_context('poster')
```

```python
gse = 'GSE54627'
fname = 'data/{}_series_matrix.txt'.format(gse)

# Comment lines begin with a variety of characters. Figure out where to start manually.
start_at = 0
for i, line in enumerate(open(fname).readlines()):
    if line.startswith('!Sample_title'):
        headers = line.split('\t')[1:]
    elif line.startswith('!Series_platform_id'):
        platform = line.split('\t')[1].strip().strip('"')
    elif line[0] not in ('!', '#', '^', '\n'):
        start_at = i
        break
data = pd.io.parsers.read_csv(fname, 
                sep='\t', header=start_at-1, index_col=0)
data = data.fillna(0)
headers = [h.strip().strip('"') for h in headers]
print(platform)
print(headers)
data.head()
```

```python
# Rename our columns
data.columns = headers
data.head()
```

```python
# Get gene names
fname = 'data/{}.txt'.format(platform)
start_at = 0
for i, line in enumerate(open(fname).readlines()):
    if line[0] not in ('!', '#', '^', '\s', '\n'):
        start_at = i
        break
probes = pd.io.parsers.read_csv(fname, 
            sep='\t', header=start_at, index_col=0)
probes.head()
```

```python
probes['Gene symbol'] = ['//' in p and p.split('//')[1].strip() or '' 
                         for p in probes['gene_assignment']]
probes.head()
```

```python
merged = pd.merge(data, probes, left_index=True, right_index=True)
merged.head()
```

```python
column_names = ('Refseq', 'value')
for sample in headers:
    subdata = data[['Refseq', sample]]
    subdata.columns = column_names
    filename = 'data/{}_{}.txt'.format(gse, sample)
    subdata.to_csv(filename,
               header=True, index=False, sep='\t')
    score = subprocess.check_output('python ./calculate_activation_signature_score.py -f ' +
                                    '{}'.format(filename), shell=True)
    print(sample, score)
```
