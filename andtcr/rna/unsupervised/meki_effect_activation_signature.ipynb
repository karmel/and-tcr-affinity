```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
sns.set_context('poster')
```

```python
# First, get PC1
filename = 'output/clusters/pca_genes_rna_and.txt'
pca = pd.io.parsers.read_csv(filename, 
                             sep='\t', 
                             header=0, index_col=0)
pca.head()
```

```python
pca = pca.sort('pc1')

ten_pct = len(pca)/10
top = pca[int(round(9*ten_pct)):]
bottom = pca[:int(ten_pct)]
len(top)
```

```python
filename = '../data/mean/rpkm.txt'
data = pd.io.parsers.read_csv(filename, 
                                    sep='\t', 
                                    header=0, index_col=0)

print(data.shape)
cols = ['No-Peptide-24h',
        '10uM-K99A-24h', 
        '100uM-K99A-24h', 
        '0-1uM-PCC-24h', 
        '10uM-PCC-24h', 
        ]
meki_cols = [s + '-MEKi' for s in cols]
labels = ['No Peptide', '10uM K99A', '100uM K99A', '0.1uM PCC', '10uM PCC']
subdata = data[cols + meki_cols]
subdata.head()
```

```python
all_cols = zip(cols, meki_cols)
all_cols = [item for sublist in all_cols for item in sublist]
```

```python
# Set up colors
colorset = sns.color_palette("husl", 5)[::-1]

# Get sets of six
color_groups = zip(colorset, colorset)

# Flatten
color_groups = [item for sublist in color_groups for item in sublist]
```

```python
# Box plots of conditions
top_rpkm = subdata.loc[top.index]
bp = plt.boxplot([top_rpkm[c] for c in all_cols], patch_artist=True)

# Colors.
for i in range(0, len(all_cols)):
    outline_col = '#666666'
    plt.setp(bp['boxes'][i], edgecolor=outline_col)
    plt.setp(bp['boxes'][i], linewidth=3)
    plt.setp(bp['boxes'][i], facecolor=color_groups[i])
    plt.setp(bp['boxes'][i], hatch=(i%2 and '//' or None))
    plt.setp(bp['medians'][i], color=outline_col)
    plt.setp(bp['whiskers'][i*2], color=outline_col)
    plt.setp(bp['whiskers'][i*2+1], color=outline_col)
    plt.setp(bp['caps'][i*2], color=outline_col)
    plt.setp(bp['caps'][i*2+1], color=outline_col)

    
# Custom legend
notx_patch = plt.Rectangle((0, 0), 1, 1, fc="white")
meki_patch = plt.Rectangle((0, 0), 1, 1, fc="white", hatch ='//')

plt.legend([notx_patch, meki_patch], ['Notx', '0.5uM MEKi'], loc='upper left')

plt.ylim([0,400])

plt.ylabel('RPKM')
plt.xticks([x + 1.5 for x in range(len(all_cols)) if x%2], ['']*len(labels))
for i, label in enumerate(labels):
    plt.text(i*2 + 1.5, -10, label,
                ha='center', va='top', fontsize=14)


title = 'MEKi effect on top 10% of PC1'
plt.title(title)

plt.savefig('figures/{}.png'.format(title.replace(' ','_')),
            bbox_inches='tight')
```

```python
# Box plots of conditions
bottom_rpkm = subdata.loc[bottom.index]
bp = plt.boxplot([bottom_rpkm[c] for c in all_cols], patch_artist=True)

# Colors.
for i in range(0, len(all_cols)):
    outline_col = '#666666'
    plt.setp(bp['boxes'][i], edgecolor=outline_col)
    plt.setp(bp['boxes'][i], linewidth=3)
    plt.setp(bp['boxes'][i], facecolor=color_groups[i])
    plt.setp(bp['boxes'][i], hatch=(i%2 and '//' or None))
    plt.setp(bp['medians'][i], color=outline_col)
    plt.setp(bp['whiskers'][i*2], color=outline_col)
    plt.setp(bp['whiskers'][i*2+1], color=outline_col)
    plt.setp(bp['caps'][i*2], color=outline_col)
    plt.setp(bp['caps'][i*2+1], color=outline_col)

    
# Custom legend
notx_patch = plt.Rectangle((0, 0), 1, 1, fc="white")
meki_patch = plt.Rectangle((0, 0), 1, 1, fc="white", hatch ='//')

plt.legend([notx_patch, meki_patch], ['Notx', '0.5uM MEKi'], loc='upper right')

plt.ylim([0,400])

plt.ylabel('RPKM')
plt.xticks([x + 1.5 for x in range(len(all_cols)) if x%2], ['']*len(labels))
for i, label in enumerate(labels):
    plt.text(i*2 + 1.5, -10, label,
                ha='center', va='top', fontsize=14)


title = 'MEKi effect on bottom 10% of PC1'
plt.title(title)

plt.savefig('figures/{}.png'.format(title.replace(' ','_')),
            bbox_inches='tight')
```

```python
for j, segment in enumerate([bottom_rpkm, top_rpkm]):
    label = 'bottom'
    if j%2: label = 'top'
    normed = segment[cols].divide(segment.max(axis=1), axis=0)
    fig = plt.figure()
    ax = fig.add_subplot(1,1,1)
    
    blues = sns.color_palette('Blues')
    normed.T.plot(legend=False, linewidth=.5, 
                  ax=ax, color=blues)

    greens = sns.color_palette('Greens')
    normed_meki = segment[meki_cols].divide(segment.max(axis=1), axis=0)
    normed_meki.T.plot(legend=False, linewidth=.5, 
                       ax=ax, color=greens)

    normed.mean().plot(legend=True, linewidth=3, 
                            ax=ax, color=sns.color_palette('Blues_d')[0],
                            label='Notx mean')
    normed_meki.mean().plot(legend=True, linewidth=3, 
                            ax=ax, color=sns.color_palette('Greens_d')[0],
                            label='MEKi mean')
    
    plt.xticks(range(0, len(labels)), labels)
    ax.set_ylabel('Normalized RPKM')
    title = 'Gene expression in the {} 10% of PC1 with MEKi'.format(label, i+1)
    ax.set_title(title)
    plt.savefig('figures/' + title.replace(' ', '_') + '.png',
        bbox_inches='tight')
```
