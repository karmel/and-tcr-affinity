```python
'''
Parse gene ontology clusters produced by David and
produce plots.
'''
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from io import StringIO
```

```python
# Get whole string
filename = 'go_analysis/pc1_top/David_enriched_clusters.txt'
cluster_str = open(filename).read()
```

```python
# Split out clusters
titled_clusters = cluster_str.split('\n\n')
```

```python
# Get enrichment score
enrichment = [float(c[c.find(': ') + 2:c.find('\n')]) for c in titled_clusters]
enrichment = pd.Series(enrichment)
```

```python
# Remove title line
text_clusters = [c[c.find('\n') + 1:] for c in titled_clusters]
```

```python
# Convert into dataframes
clusters = [pd.io.parsers.read_csv(StringIO(c), sep='\t', header=0) for c in text_clusters]
```

```python
clusters[0].head()
```

```python
# Make a single dataframe with the first row of each cluster
top_matches = [c.loc[0].tolist() for c in clusters]
matches = pd.DataFrame(top_matches, columns=clusters[0].columns)
```

```python
# Clean names
matches['Ontology term'] = matches['Term'].str.replace(r'GO:.*~','')
matches
```

```python
# Some redundancy. Take interesting sets.
subdata = matches.loc[list(range(8)) + [16,17]]
```

```python
# Plot p-val, with number of genes as size of circle.
scalor = 10
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.scatter(subdata['PValue'], enrichment.loc[subdata.index],
            s=subdata['Count']*scalor, alpha=.5)
ax.set_xscale('log')
ax.set_xlim([0, 1e-30])

# Add labels. Shift some of text so that it's legible.
text_loc = enrichment.loc[subdata.index].tolist()
text_loc[2] -= .6
text_loc[3] += .2
text_loc[4] += .2
text_loc[8] += .8 
for i, label in enumerate(subdata['Ontology term']):
    ax.text(subdata['PValue'].loc[subdata.index[i]]/3, 
            text_loc[i],
            label, fontsize=16)


plt.ylabel('DAVID Cluster Enrichment Score')
plt.xlabel('Term P-Value')
title = 'Gene Ontology Enrichment for Activation Signature Genes'
plt.title(title, y=1.01)
plt.savefig('figures/{}'.format(title.replace(' ','_')),
            bbox_inches='tight')
plt.show()
```

```python
# And just in chart form.
subdata['p-value'] = ['%.2g' % s for s in subdata['PValue']]
chartable = subdata[['Ontology term', 'p-value', 'Count']]
chartable.index = range(1, len(chartable)+1)
chartable
```
