```python
'''
We want to cluster the genes in the top/bottom of each to find a group
that might make good biomarkers for AP-1 responsiveness (PC1)
and low versus high affinity signaling (PC2).
'''
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import scale
```

```python
sns.set_context('poster')
```

```python
filename = 'output/clusters/pca_genes_rna_and.txt'
pca = pd.io.parsers.read_csv(filename, 
                             sep='\t', 
                             header=0, index_col=0)
cols = ['No-Peptide-24h',
        '10uM-K99A-24h', 
        '100uM-K99A-24h', 
        '0-1uM-PCC-24h', 
        '10uM-PCC-24h', 
        ]
pep_names = ['No Peptide', '10uM K99A', '100uM K99A', '0.1uM PCC', '10uM PCC']
pca.head()
```

```python
k = 8
models = []
datasets = []
scaled_data = []
for x in (1, 2):
    pca_sorted = pca.sort('pc' + str(x))
    ten_pct = len(pca_sorted)//10
    bottom = pca_sorted[:ten_pct]
    top = pca_sorted[9*ten_pct:]
    datasets.append(bottom)
    datasets.append(top)
    for segment in (bottom, top):
        scaled = segment[cols].divide(segment[cols].max(axis=1), axis=0)
        scaled_data.append(scaled)
        model = KMeans(n_clusters=k)
        model.fit(scaled)
        models.append(model)
```

```python
segments = ['PC1 Bottom 10%',
          'PC1 Top 10%',
          'PC2 Bottom 10%',
          'PC2 Top 10%',]

for i, segment in enumerate(segments):
    model = models[i]
    data = datasets[i]
    
    # Add labels for later output
    data['kmeans'] = model.labels_
    
    # Make a dataframe with our scaled data
    scaled = scaled_data[i]
    scaled_df = pd.DataFrame(scaled, index=data.index)
    # Sort according to our labels
    scaled_df['kmeans'] = model.labels_

    clustered = scaled_df.sort(columns=['kmeans'] + cols)
    
    data_sorted = data.loc[clustered.index]
    #if i == 1: print(vals_sorted.loc['NM_013693'])
    
    plt.figure(figsize=(15, 50))
    
    plt.pcolor(clustered[cols], cmap='coolwarm')
    
    xcol_loc = [x + .5 for x in range(0, len(pep_names))]
    plt.xticks(xcol_loc, pep_names)
    ycol_loc = [x + .5 for x in range(0, len(clustered))]
    plt.yticks(ycol_loc, data_sorted['gene_name'], size=10)
    
    plt.ylim([0,len(clustered)])
    
    plt.colorbar()
    title = 'K-means clustering of {}, k={}'.format(segment, k)
    plt.title(title)
    plt.savefig('figures/{}.png'.format(title.replace(' ','_')))
    
    data_sorted.to_csv('output/clusters/{}.txt'.format(title.replace(' ','_')),
                       sep='\t', header=True, index=True,)
```
