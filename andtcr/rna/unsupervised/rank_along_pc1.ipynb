```python
'''
Given a dataset with RPKM values, assign a numerical score
along PC1. This score serves as the relative activation status
of the sample.
'''
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
sns.set_context('poster')
```

```python
# First, get PC1
filename = 'output/clusters/pca_genes_rna_and.txt'
pca = pd.io.parsers.read_csv(filename, 
                             sep='\t', 
                             header=0, index_col=0)
pca.head()
```

```python
pc1_vector = pca['pc1'].copy()
pc1_vector.sort()

ten_pct = len(pc1_vector)/10
act_signature = pc1_vector[int(round(9*ten_pct)):]
len(act_signature)
```

```python
filename = '../data/mean/rpkm.txt'
data = pd.io.parsers.read_csv(filename, 
                                    sep='\t', 
                                    header=0, index_col=0)

print(data.shape)
cols = ['No-Peptide-24h',
        '10uM-K99A-24h', 
        '100uM-K99A-24h', 
        '0-1uM-PCC-24h', 
        '10uM-PCC-24h', 
        ]
meki_cols = [s + '-MEKi' for s in cols]
labels = ['No Peptide', '10uM K99A', '100uM K99A', '0.1uM PCC', '10uM PCC']
subdata = data[cols]
subdata.head()
```

```python
for col in cols:
    transformed = np.dot(data[col].loc[act_signature.index], act_signature)
    print(col, transformed)
```

```python
for col in meki_cols:
    transformed = np.dot(data[col].loc[act_signature.index], act_signature)
    print(col, transformed)
```

```python
def norm(column):
    # Normalize based on total tags/expression
    scale = 1e4/sum(column)
    return column*scale
```

```python
scores = []
for pair in zip(cols, meki_cols):
    for col in pair:
        normed = norm(data[col])
        transformed = np.dot(normed.loc[act_signature.index], act_signature)
        scores.append(transformed)
scores
```

```python
# Set up colors
colors = sns.color_palette("husl", 5)[::-1]

# Get sets of two
triples = zip(colors, colors)

# Flatten
triples = [item for sublist in triples for item in sublist]
```

```python
x_range = range(0, len(scores))
plt.figure()

bars = plt.bar([x + .2 for x in x_range], scores, color=list(triples))

# Add hatching on every other bar
for i, bar in enumerate(bars):
    if i%2 == 1:
        bar.set_hatch('//')
    else:
        bar.set_hatch(None)

# Custom legend
notx_patch = plt.Rectangle((0, 0), 1, 1, fc="white")
meki_patch = plt.Rectangle((0, 0), 1, 1, fc="white", hatch ='//')

plt.legend([notx_patch, meki_patch], ['Notx', '0.5uM MEKi'], loc='upper left')

# Label scores above bars
for i, score in enumerate(scores):
    plt.text(i + .6, 1.01*score, '%.2f' % score,
                ha='center', va='bottom', fontsize=14)

plt.xlim([0, len(scores) + .2])
plt.ylabel('Activation Signature Score')

plt.xticks([x + 1.1 for x in x_range if not x%2], labels)

title = 'Activation Signature Scores with MEKi Treatment'
plt.title(title)
plt.savefig('figures/{}.png'.format(title.replace(' ', '_')), bbox_inches='tight')
```
