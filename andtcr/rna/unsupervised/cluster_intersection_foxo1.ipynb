```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set_context('poster')
```

```python
# Sql was run on DB from separate client. 
# Can be run from here with SqlAlchemy if desired.
# Generates table with transcribed genes, 
# outer joined to grab foxo1 peaks in tss +- 1000bp or within 1000bp of gene body.
# Only one row per refseq identifier, so multiple peaks in a single gene are not 
# captured here.
query = '''select distinct on (seq.sequence_identifier)
t.id, seq.sequence_identifier, chr.name as chr_name, t.transcription_start, t.transcription_end, t.strand,
t.score, t.rpkm, 
foxo_naive_gene.id as gene_peak_id, 
foxo_naive_gene.start as gene_peak_start, 
foxo_naive_gene."end" as gene_peak_end, 
foxo_naive_gene.tag_count as gene_peak_tag_count,
foxo_naive_tss.id as tss_peak_id, 
foxo_naive_tss.start as tss_peak_start, 
foxo_naive_tss."end" as tss_peak_end, 
foxo_naive_tss.tag_count as tss_peak_tag_count

from atlas_mm10_default.atlas_transcript t
join genome_reference_mm10.chromosome chr
on t.chromosome_id = chr.id
join atlas_mm10_default.atlas_transcript_sequence tseq
on t.id = tseq.atlas_transcript_id
join genome_reference_mm10.sequence_transcription_region reg
on tseq.sequence_transcription_region_id = reg.id
join genome_reference_mm10.sequence_identifier seq
on reg.sequence_identifier_id = seq.id

left outer join chipseq.peak_cd8tcell_foxo1_naive foxo_naive_gene
on reg.chromosome_id = foxo_naive_gene.chromosome_id
and reg.start_end_1000 && foxo_naive_gene.start_end
and foxo_naive_gene.tag_count >= 20

left outer join chipseq.peak_cd8tcell_foxo1_naive foxo_naive_tss
on reg.chromosome_id = foxo_naive_tss.chromosome_id
and reg.start_site_1000 && foxo_naive_tss.start_end
and foxo_naive_tss.tag_count >= 20

where t.refseq = true
and tseq.major = true
'''
```

```python
gene_peaks = '../data/genes_with_foxo1.txt'
gene_peaks = pd.io.parsers.read_csv(gene_peaks, 
                                    sep='\t', 
                                    header=0, index_col=1)

rpkm = '../data/rpkm.txt'
rpkm = pd.io.parsers.read_csv(rpkm, 
                                    sep='\t', 
                                    header=0, index_col=0)

clusters = 'output/affinity_pairwise_wt_ko.txt'
clusters = pd.io.parsers.read_csv(clusters, 
                                    sep='\t', 
                                    header=0, index_col=0)
clusters.head()
```

```python
# Isolate transcribed genes.
transcribed = pd.merge(gene_peaks, rpkm, how='inner', left_index=True, right_index=True)
transcribed.head()
```

```python
len(transcribed), len(gene_peaks)
```

```python
# What percent of transcribed genes have Foxo1 peaks?
counts = pd.Series([sum(transcribed['gene_peak_id'] > 0), 
                   sum(transcribed['gene_peak_id'].isnull())],
                   index=['Has Foxo1', 'No Foxo1'])
counts.plot(kind='pie', colors=sns.color_palette('coolwarm', 3),
            autopct='%.2f', fontsize=20, figsize=(15, 15))
title = 'Expressed genes with Foxo1 binding within 1000 bp'
plt.title(title + '\n(Total genes: {})'.format(len(transcribed)), fontweight='bold')
plt.savefig('figures/{}.png'.format(title.replace(' ','_')))
```

```python
# What percent of differential genes have Foxo1 peaks?
differential = pd.merge(clusters, transcribed, how='inner', left_index=True, right_index=True)
counts = pd.Series([sum(differential['gene_peak_id'] > 0), 
                   sum(differential['gene_peak_id'].isnull())],
                   index=['Has Foxo1', 'No Foxo1'])
counts.plot(kind='pie', colors=sns.color_palette('coolwarm', 3),
            autopct='%.2f', fontsize=20, figsize=(15, 15))
```

```python
# For each cluster?
labels = clusters['label'].unique()
for x in labels:
    # Arrange the subplots in reverse
    plt.subplot(410 + (len(labels) - x))
    clust = clusters[clusters['label'] == x]
    diff_clust = pd.merge(clust, transcribed, how='inner', left_index=True, right_index=True)
    counts = pd.Series([sum(diff_clust['gene_peak_id'] > 0), 
                       sum(diff_clust['gene_peak_id'].isnull())],
                       index=['Has Foxo1', 'No Foxo1'])
    counts.plot(kind='pie', colors=sns.color_palette('coolwarm', 3),
                autopct='%.2f', fontsize=16, figsize=(4, 18))
    plt.title('Cluster {}'.format(x + 1), fontweight='bold')

plt.savefig('figures/foxo1_binding_by_cluster.png', bbox_inches='tight')
```

```python
# Same, but for TSS?
labels = clusters['label'].unique()
for x in labels:
    # Arrange the subplots in reverse
    plt.subplot(410 + (len(labels) - x))
    clust = clusters[clusters['label'] == x]
    diff_clust = pd.merge(clust, transcribed, how='inner', left_index=True, right_index=True)
    counts = pd.Series([sum(diff_clust['tss_peak_id'] > 0), 
                       sum(diff_clust['tss_peak_id'].isnull())],
                       index=['Has Foxo1', 'No Foxo1'])
    counts.plot(kind='pie', colors=sns.color_palette('coolwarm', 3),
                autopct='%.2f', fontsize=16, figsize=(4, 18))
    plt.title('Cluster {}'.format(x + 1), fontweight='bold')
```
