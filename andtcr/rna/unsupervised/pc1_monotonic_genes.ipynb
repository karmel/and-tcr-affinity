```python
'''
We are interested in exploring the genes that are monotonically 
increasing in the top of PC1, and monotonically decreasing
in the bottom.

First, find the genes of interest.
'''
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
sns.set_context('poster')
```

```python
def is_monotonic(row, columns, increase=0):
    ''' 
    For a given row, is it monotonically increasing?
    Not that if we swap the order of the input columns,
    we are measuring decreasing instead.
    
    `increase` is the percent increase required;
    a value of 0 indicates any amount greater counts,
    whereas 0.05 would require more than a 5% increase.
    '''
    last = -np.inf
    for c in columns:
        if row[c]*(1 + increase) <= last:
            return False
        last = row[c]
    return True
```

```python
cols = ['No-Peptide-24h',
        '10uM-K99A-24h', 
        '100uM-K99A-24h', 
        '0-1uM-PCC-24h', 
        '10uM-PCC-24h', 
        ]
pep_names = ['No Peptide', '10uM K99A', '100uM K99A', '0.1uM PCC', '10uM PCC']
```

```python
# For PC1 Top
segment = 'top'
direction = 'increasing'
sort_cols = cols
filename = 'output/clusters/pca_pc1_{}_rna_and.txt'.format(segment)
data = pd.io.parsers.read_csv(filename, 
                             sep='\t', 
                             header=0, index_col=0)
data.head()
```

```python
monotonic = data[data.apply(lambda x: is_monotonic(x, sort_cols), axis=1)]
monotonic.head()
```

```python
scaled = monotonic[cols].divide(monotonic[cols].max(axis=1), axis=0)
ordered = scaled.sort(sort_cols)

data_ordered = monotonic.loc[ordered.index]

plt.figure(figsize=(15, 20))


plt.pcolor(ordered, cmap='coolwarm')

xcol_loc = [x + .5 for x in range(0, len(pep_names))]
plt.xticks(xcol_loc, pep_names)
ycol_loc = [x + .5 for x in range(0, len(ordered))]
plt.yticks(ycol_loc, data_ordered['gene_name'], size=10)

plt.ylim([0,len(ordered)])

plt.colorbar()
title = 'Monotonically {} genes in the {} 10% of PC1'.format(direction, segment)
plt.title(title)
plt.savefig('figures/{}.png'.format(title.replace(' ','_')))

data_ordered.to_csv('output/clusters/monotonic_pc1_{}_{}.txt'.format(segment, direction),
                   sep='\t', header=True, index=True,)
```

```python
# For PC1 Top
segment = 'bottom'
direction = 'decreasing'
sort_cols = cols[::-1]
filename = 'output/clusters/pca_pc1_{}_rna_and.txt'.format(segment)
data = pd.io.parsers.read_csv(filename, 
                             sep='\t', 
                             header=0, index_col=0)
data.head()
```
