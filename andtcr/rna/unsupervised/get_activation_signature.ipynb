```python
'''
We want the intersection of all three activation signature gene sets--
replicate 1, replicate 2, and the mean.
'''
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
sns.set_context('poster')
```

```python
# First, get PC1
filename = 'output/clusters/pca_genes_rna_and_rep1.txt'
pca1 = pd.io.parsers.read_csv(filename, 
                             sep='\t', 
                             header=0, index_col=0)
filename = 'output/clusters/pca_genes_rna_and_rep2.txt'
pca2 = pd.io.parsers.read_csv(filename, 
                             sep='\t', 
                             header=0, index_col=0)

filename = 'output/clusters/pca_genes_rna_and.txt'
pca_both = pd.io.parsers.read_csv(filename, 
                             sep='\t', 
                             header=0, index_col=0)
pca_both.head()
```

```python
# Sort.
pca1 = pca1.sort('pc1')

fraction = 10
ten_pct = len(pca1)/fraction
top1 = pca1[int(round((fraction-1)*ten_pct)):]
bottom1 = pca1[:int(ten_pct)]
print(len(top1))

pca2 = pca2.sort('pc1')

ten_pct = len(pca2)/fraction
top2 = pca2[int(round((fraction-1)*ten_pct)):]
bottom2 = pca2[:int(ten_pct)]
print(len(top2))

pca_both = pca_both.sort('pc1')
ten_pct = len(pca_both)/fraction
top_both = pca_both[int(round((fraction-1)*ten_pct)):]
bottom_both = pca_both[:int(ten_pct)]
print(len(top_both))
```

```python
# Throw out genes where no-pep standard deviation is too high
cutoff = .2
min_exp = 100
top_both = top_both[(top_both['No-Peptide-24h'] <= min_exp) |
    (top_both['No-Peptide-24h']*cutoff > top_both['No-Peptide-24h-std'])]
bottom_both = bottom_both[(bottom_both['No-Peptide-24h'] <= min_exp) |
    (bottom_both['No-Peptide-24h']*cutoff > bottom_both['No-Peptide-24h-std'])]
print(len(top_both))
print(len(bottom_both))
```

```python
# Get intersection of all three sets.
intersect_top = set(top1.index.tolist()) &\
            set(top2.index.tolist()) &\
            set(top_both.index.tolist())
print(len(intersect_top))
```

```python
# Get intersection of all three sets.
intersect_bottom = set(bottom1.index.tolist()) &\
            set(bottom2.index.tolist()) &\
            set(bottom_both.index.tolist())
print(len(intersect_bottom))
```

```python
pca_intersect = pca_both.loc[list(intersect_top) + list(intersect_bottom)]
pca_intersect = pca_intersect.sort('pc1')
act_signature = pca_intersect[['gene_name', 'pc1']]
act_signature['Refseq'] = act_signature.index
act_signature.columns = ['gene symbol', 'pc1', 'refseq']
act_signature = act_signature[['refseq', 'gene symbol', 'pc1']]
act_signature.head()
```

```python
act_signature.to_csv('../activationscore/activation_signature.txt',
                     header=True, index=False, sep='\t')
```
