```python
'''
Are the groupings seen in the top 10% and bottom 10% significant?

Do permutation testing-- for groups of the same size, calculate mean,
then calculate difference between no peptide and 10uM PCC at the mean line.

Determine the distribution of the mean-diff statistic, and then compare
our actual group mean-diffs wising the appropriate test.
'''
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
filename = 'output/clusters/pca_genes_rna_and.txt'
pca = pd.io.parsers.read_csv(filename, 
                             sep='\t', 
                             header=0)
pca.describe()
```

```python
cols = ['No-Peptide-24h',
        '10uM-K99A-24h', 
        '100uM-K99A-24h', 
        '0-1uM-PCC-24h', 
        '10uM-PCC-24h', 
        ]
labels = ['No Peptide', '10uM K99A', '100uM K99A', '0.1uM PCC', '10uM PCC']
```

```python
def get_mean_diff(df):
    # Get mean.
    df_sub = df[cols]
    avg = df_sub.mean(axis=0)
    return avg[cols[-1]] - avg[cols[0]]
```

```python
# Generate our distribution
ten_pct = round(len(pca)/10)
mean_diffs = []
for _ in range(10000):
    rand_rows = np.random.random_integers(0, len(pca), ten_pct)
    subdata = pca.loc[rand_rows]
    mean_diffs.append(get_mean_diff(subdata))
```

```python
plt.hist(mean_diffs)
```

```python
# Get mean_diff for groups of interest
pc1_sorted = pca.sort('pc1')
pc1_bottom = pc1_sorted[:int(ten_pct)]
pc1_top = pc1_sorted[int(9*ten_pct):]
bottom_diff = get_mean_diff(pc1_bottom)
top_diff = get_mean_diff(pc1_top)
print(bottom_diff)
print(top_diff)
```

```python
# Using the actual distribution gives us a p-value of 0. 
# That distribution plot looks mostly normal, so use normal distribution.
avg_diff = np.mean(mean_diffs)
std_diff = np.std(mean_diffs)
print (avg_diff, std_diff)
```

```python
from scipy.stats import norm
rv = norm(avg_diff, std_diff)
2*rv.cdf(bottom_diff)
```

```python
2*(1 - rv.cdf(top_diff))
```
