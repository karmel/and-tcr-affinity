```python
'''
GSE49929: OT-1 CD8 T cells responding to N4 (high) or V4 (low)
'''
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set_context('poster')
```

```python
gse = 'GSE49929'
fname = 'data/{}_cd8tcell_irf4_wt_rpkm.txt'.format(gse)
data = pd.io.parsers.read_csv(fname, 
                sep='\t', header=0, index_col=0)
data = data.fillna(0)
data.head()
```

```python
# Get sample names
headers = data.filter(like='FPKM').columns
headers
```

```python
high, low, notx = [], [], []
for header in headers:
    if 'N4' in header: high.append(header)
    elif 'V4' in header: low.append(header)
    elif 'notx' in header: notx.append(header)
```

```python
genes_of_interest = ['Cd83', 'Gata3', 'Ikzf2', 'Kit',
                     'Gpr183', 'Lad1', 'Tnfsf8', 
                     'Ccl4', 'Cd5', 'Rgs16', 'Nr4a2']
for gene in genes_of_interest:
    gene_data = data[data['Annotation/Divergence'].str.contains(gene)]
    
    # Reorder columns
    gene_groups = [gene_data[notx], gene_data[low], gene_data[high]]
    
    plt.figure()
    plt.boxplot(gene_groups)

    
    plt.ylabel('RPKM')
    
    labels = ['Notx (OT-1)', 'Low affinity (OT-1 + V4)', 'High affinity (OT-1 + N4)']
    plt.xticks(range(1,len(labels)+1),
               labels)

    title = '{}- OT-1 CD8 T cells responding to N4 or V4- {}'.format(gse, gene)
    plt.title(title)

    plt.savefig('figures/{}.png'.format(title.replace(' ','_')),
                bbox_inches='tight')
```
