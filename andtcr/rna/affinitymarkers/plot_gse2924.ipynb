```python
'''
GSE2924: Immunodominance of CD8 T cells.
H7a is high affinity, HY is low affinity, plus no treatment.
'''
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set_context('poster')
```

```python
gse = 'GSE2924'
fname = 'data/{}_series_matrix.txt'.format(gse.lower())
data = pd.io.parsers.read_csv(fname, 
                sep='\t', header=0, index_col=0,
                comment='!')
data = data.fillna(0)
data.head()
```

```python
# Get sample names
for line in open(fname).readlines():
    if line.startswith('!Sample_title'):
        headers = line.split('\t')[1:]
    elif line.startswith('!Series_platform_id'):
        platform = line.split('\t')[1].strip().strip('"')
    elif line.startswith('!series_matrix_table_begin'):
        break
headers = [h.strip().strip('"') for h in headers]
print(platform)
headers
```

```python
high, low, notx = [], [], []
for header in headers:
    if 'H7a' in header: high.append(header)
    elif 'HY' in header: low.append(header)
    else: notx.append(header)
```

```python
# Rename our columns
data.columns = headers
data.head()
```

```python
# Get gene names
# Ugh. Platform file has three different comment chars in the
# header lines. Hardcode header for now...
fname = 'data/{}.txt'.format(platform)
probes = pd.io.parsers.read_csv(fname, 
                sep='\t', header=18, index_col=0)
probes.head()
```

```python
genes_of_interest = ['Cd83', 'Gata3', 'Ikzf2', 'Kit',
                     'Gpr183', 'Lad1', 'Tnfsf8', 
                     'Ccl4', 'Cd5', 'Rgs16', 'Nr4a2']

for gene in genes_of_interest:
    probe_set = probes[probes['Gene Symbol'] == gene].index
    
    # Get rows for probes of interest.
    gene_data = data.loc[probe_set]
    
    # Pick the row with the highest mean intensity
    mean_intensity = gene_data.mean(axis=1)
    mean_intensity.sort(ascending=False)
    
    probe_data = gene_data.loc[mean_intensity.index[0]]
    # Reorder columns
    probe_groups = [probe_data[notx], probe_data[low], probe_data[high]]
    
    plt.figure()
    plt.boxplot(probe_groups)

    
    plt.ylabel('Probe Intensity')
    labels = ['Notx (CD8)', 'Low affinity (CD8 + HY)', 'High affinity (CD8 + H7a)']
    plt.xticks(range(1,len(labels)+1),
               labels)

    title = '{}- Immunodominance in CD8 T cell responses- {}'.format(gse,gene)
    plt.title(title)

    plt.savefig('figures/{}.png'.format(title.replace(' ','_')),
                bbox_inches='tight')
```
