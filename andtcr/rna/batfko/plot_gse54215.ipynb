```python
'''
GSE54215: Microarray of Batf KO CD8s.
Is PC1 preferentially downregulated?

Post-script: preferentially upregulated instead. This is in line
with what the authors said in the paper (Ifng, other effectors are up),
where it was attributed to dysregulation of a network of transcription
factors. Hmm. That makes things more complicated.
'''
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
sns.set_context('poster')
```

```python
gse = 'GSE54215'
fname = 'data/{}_series_matrix.txt'.format(gse)
data = pd.io.parsers.read_csv(fname, 
                sep='\t', header=0, index_col=0,
                comment='!')
data = data.fillna(0)
data.head()
```

```python
# Get sample names
for line in open(fname).readlines():
    if line.startswith('!Sample_title'):
        headers = line.split('\t')[1:]
    elif line.startswith('!Series_platform_id'):
        platform = line.split('\t')[1].strip().strip('"')
    elif line.startswith('!series_matrix_table_begin'):
        break
headers = [h.strip().strip('"') for h in headers]
print(platform)
headers
```

```python
wt, ko = [], []
for header in headers:
    if 'WT_P14_D3_Effector' in header: wt.append(header)
    elif 'BATF_KO_P14_D3_Effector' in header: ko.append(header)
```

```python
# Rename our columns
data.columns = headers
data.head()
```

```python
# Get gene names
# Ugh. Platform file has three different comment chars in the
# header lines. Hardcode header for now...
fname = 'data/{}.txt'.format(platform)
probes = pd.io.parsers.read_csv(fname, 
                sep='\t', header=18)
probes.head()
```

```python
# Get the PC1 top and bottom genes.
segment = 'top'
filename = '../unsupervised/output/clusters/pca_pc1_{}_rna_and.txt'.format(segment)
top = pd.io.parsers.read_csv(filename, 
                             sep='\t', 
                             header=0, index_col=0)
segment = 'bottom'
filename = '../unsupervised/output/clusters/pca_pc1_{}_rna_and.txt'.format(segment)
bottom = pd.io.parsers.read_csv(filename, 
                             sep='\t', 
                             header=0, index_col=0)
bottom.head()
```

```python
# Get probe ids of interest.
top_merge = pd.merge(probes, top,
                      how='inner', left_on='Gene Symbol', right_on='gene_name')
top_probes = top_merge['ID']
print(top_probes.head())


bottom_merge = pd.merge(probes, bottom,
                      how='inner', left_on='Gene Symbol', right_on='gene_name')
bottom_probes = bottom_merge['ID']
bottom_probes.head()
```

```python
# Calculate log fold change for each probe in our data.
data['wt_mean'] = data[wt].mean(axis=1)
data['ko_mean'] = data[ko].mean(axis=1)
data['fc'] = data['ko_mean']/data['wt_mean']
data['log_fc'] = np.log2(data['fc'])
data.head()
```

```python
# Plot logfc for the top and bottom genes.

plt.figure()

data_groups = [data['log_fc'],
                data.loc[bottom_probes]['log_fc'],
                data.loc[top_probes]['log_fc']]

plt.boxplot(data_groups)


plt.ylabel('Log Fold Change (KO/WT)')

labels = ['All genes', 'PC1 bottom 10%', 'PC1 top 10%']
plt.xticks(range(1,len(labels)+1),
           labels)
plt.ylim([-1, 1])
title = '{}- Batf KO effect on PC1 segments'.format(gse)
plt.title(title)

plt.savefig('figures/{}.png'.format(title.replace(' ','_')),
            bbox_inches='tight')
```
