```python
'''
For any given gene that is changing with KLA treatment, 
find the nearest neighbor non-genic transcripts.

Do we want to separate by chromosome here? Maybe not;
trans effects can in theory be captured, and might
be a good measure for how precise our neighborhood is,
given expected rates of trans vs. cis interactions.
'''
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.neighbors import NearestNeighbors
from sklearn.preprocessing import scale
```

```python
sns.set_context('poster')
```

```python
genes = pd.io.parsers.read_csv('data/refseq_genes_kla.txt', 
                                    sep='\t', 
                                    header=0, index_col=0)
nongenes = pd.io.parsers.read_csv('data/nongenes_kla.txt', 
                                    sep='\t', 
                                    header=0, index_col=0)
```

```python
# First, normalize the tag count data
norm_tags = 1e7
for data in (genes, nongenes):
    for col in data.filter(like='tag_count').columns:
        prefix = col.replace('_tag_count','')
        data[prefix + '_norm'] = data[col] * (norm_tags/data[prefix + '_total_tags'])
        
```

```python
# We only care about changing genes.
fold = 2
genes_mask = genes.filter(like='_norm').min(axis=1)*fold <= genes.filter(like='_norm').max(axis=1)
genes_chg = genes[genes_mask]
genes_chg.head()
```

```python
# Extract columns of interest and scale row-wise to unit norm.
genes_v = genes_chg.filter(like='_norm')
nongenes_v = nongenes.filter(like='_norm')

for data in (genes_v, nongenes_v):
    max_rowwise = data.max(axis=1)
    for col in data.columns:
        data[col] = data[col]/max_rowwise
    
genes_v.head()
```

```python
# Add back in chromosome column, and scale so that it has greater weight.
chr_weight = 1.0
genes_v['chr'] = genes_chg['chromosome_id'] * chr_weight
nongenes_v['chr'] = nongenes['chromosome_id'] * chr_weight

# Add back in transcription start and end, and scale across all samples
bp_weight = 10
genes_v['start'] = genes_chg['transcription_start']/genes_chg['transcription_start'].max()*bp_weight
genes_v['end'] = genes_chg['transcription_end']/genes_chg['transcription_end'].max()*bp_weight
nongenes_v['start'] = nongenes['transcription_start']/nongenes['transcription_start'].max()*bp_weight
nongenes_v['end'] = nongenes['transcription_end']/nongenes['transcription_end'].max()*bp_weight
```

```python
# We want to set a radius rather than number of neighbors,
# as not all genes will have the same number of "enhancers"
radii = np.arange(0,.5,.025)
              

nbrs = NearestNeighbors(algorithm='kd_tree').fit(nongenes_v)
one_nbr = []
for i, r in enumerate(radii):
    distances, indices = nbrs.radius_neighbors(genes_v, radius=r)
    
    # Draw histogram of number of neighbors found.
    counts = [len(ind_set) for ind_set in indices]
    cols = 4
    ax = plt.subplot(np.ceil(len(radii)/cols), cols, i + 1)
    plt.hist(counts)
    plt.title('Radius = {}'.format(r), fontsize=10)
    ax.tick_params(axis='x', labelsize=10)
    ax.tick_params(axis='y', labelsize=10)

    # How many genes have exactly one neighbor?
    one_nbr.append(len([x for x in counts if x == 1]))
plt.tight_layout()
    
plt.figure()
plt.plot(radii, one_nbr, 'o-')
```

```python
# HiC data is our True label. Evaluate rate of retrieving True interactions.
radii = np.arange(0,.5,.025)
              
matrix = [] # @todo
resolution = 10000

nbrs = NearestNeighbors(algorithm='kd_tree').fit(nongenes_v)
mean_rates = []
for i, r in enumerate(radii):
    distances, indices = nbrs.radius_neighbors(genes_v, radius=r)
    
    ixn_rates = []
    for (idx, gene), ind_set in zip(genes_v.iterrows(), indices):
        # Find the gene in the interaction matrix
        start = genes_chg.loc[idx]['transcription_start']//resolution * resolution
        chrom = genes_chg.loc[idx]['chr_name']
        gene_column = matrix[chrom + '-' + start]
        
        # For each predicted partner, what is our interaction expectation?
        ng_indices = nongenes_v.index[ind_set]
        for ind in ng_indices:
            start = nongenes.loc[ind]['transcription_start']//resolution * resolution
            chrom = nongenes.loc[ind]['chr_name']
            ixn_rates.append(gene_column.loc[chrom + '-' + start].tolist()[0])
            
    # Draw histogram of interaction rates from HiC for predicted partners.
    cols = 4
    ax = plt.subplot(np.ceil(len(radii)/cols), cols, i + 1)
    plt.hist(ixn_rates)
    plt.title('Radius = {}'.format(r), fontsize=10)
    ax.tick_params(axis='x', labelsize=10)
    ax.tick_params(axis='y', labelsize=10)

    # What is the mean
    mean_rates.append(np.mean(ixn_rates))
plt.tight_layout()
    
plt.figure()
plt.plot(radii, mean_rates, 'o-')
```
