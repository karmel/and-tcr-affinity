```python
from pandas.io.sql import read_sql_query
from sqlalchemy import create_engine


def get_engine(uri, password, username='vespucci_user', dbname='vespucci'):
    '''
    dialect+driver://username:password@host:port/database
    '''
    return create_engine('postgresql://{}:{}@{}:5432/{}'.format(
        username, password, uri, dbname))


def dataframe_from_query(query, engine):
    '''
    Given the database connection and query, return a Pandas dataframe
    with the requested data.
    '''
    return read_sql_query(query, engine)
```

```python
engine = get_engine(
    uri='ec2-54-160-104-39.compute-1.amazonaws.com',
    password='vespucci 2015')
```

```python
gene_sql = '''
SELECT
DISTINCT ON(t.id)
t.id,
chr.name as chr_name,
t.transcription_start,
t.transcription_end,
t.strand,
t.score,
t.rpkm,
t.chromosome_id,
seq.sequence_identifier,
notx.tag_count as notx_tag_count,
kla_1h.tag_count as kla_1h_tag_count,
kla_6h.tag_count as kla_6h_tag_count,
kla_12h.tag_count as kla_12h_tag_count,
kla_24h.tag_count as kla_24h_tag_count,
notx_run.total_tags as notx_total_tags,
kla_1h_run.total_tags as kla_1h_total_tags,
kla_6h_run.total_tags as kla_6h_total_tags,
kla_12h_run.total_tags as kla_12h_total_tags,
kla_24h_run.total_tags as kla_24h_total_tags

FROM atlas_mm10_default.atlas_transcript t
join genome_reference_mm10.chromosome chr
on t.chromosome_id = chr.id

-- Gene region and name
join atlas_mm10_default.atlas_transcript_sequence tseq
on t.id = tseq.atlas_transcript_id
join genome_reference_mm10.sequence_transcription_region reg
on tseq.sequence_transcription_region_id = reg.id
join genome_reference_mm10.sequence_identifier seq
on reg.sequence_identifier_id = seq.id

-- Tag counts
join atlas_mm10_default.atlas_transcript_source notx
on t.id = notx.atlas_transcript_id
join genome_reference_mm10.sequencing_run notx_run
on notx.sequencing_run_id = notx_run.id
and notx_run.source_table = 'groseq"."tag_thiomac_gro_notx_1'

join atlas_mm10_default.atlas_transcript_source kla_1h
on t.id = kla_1h.atlas_transcript_id
join genome_reference_mm10.sequencing_run kla_1h_run
on kla_1h.sequencing_run_id = kla_1h_run.id
and kla_1h_run.source_table = 'groseq"."tag_thiomac_gro_kla_1h_1'

join atlas_mm10_default.atlas_transcript_source kla_6h
on t.id = kla_6h.atlas_transcript_id
join genome_reference_mm10.sequencing_run kla_6h_run
on kla_6h.sequencing_run_id = kla_6h_run.id
and kla_6h_run.source_table = 'groseq"."tag_thiomac_gro_kla_6h_1'

join atlas_mm10_default.atlas_transcript_source kla_12h
on t.id = kla_12h.atlas_transcript_id
join genome_reference_mm10.sequencing_run kla_12h_run
on kla_12h.sequencing_run_id = kla_12h_run.id
and kla_12h_run.source_table = 'groseq"."tag_thiomac_gro_kla_12h_1'

join atlas_mm10_default.atlas_transcript_source kla_24h
on t.id = kla_24h.atlas_transcript_id
join genome_reference_mm10.sequencing_run kla_24h_run
on kla_24h.sequencing_run_id = kla_24h_run.id
and kla_24h_run.source_table = 'groseq"."tag_thiomac_gro_kla_24h_1'

where
t.refseq = true
and tseq.major = true
and t.score >= 1
'''
```

```python
data = dataframe_from_query(gene_sql, engine)

data.to_csv('data/refseq_genes_kla.txt', sep='\t', header=True, index=False)
```
