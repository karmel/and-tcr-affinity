```python
from collections import OrderedDict
import pandas as pd
```

```python
from pandas.io.sql import read_sql_query
from sqlalchemy import create_engine


def get_engine(uri, password, username='vespucci_user', dbname='vespucci'):
    '''
    dialect+driver://username:password@host:port/database
    '''
    return create_engine('postgresql://{}:{}@{}:5432/{}'.format(
        username, password, uri, dbname))


def dataframe_from_query(query, engine):
    '''
    Given the database connection and query, return a Pandas dataframe
    with the requested data.
    '''
    return read_sql_query(query, engine)
```

```python
engine = get_engine(
    uri='ec2-54-160-104-39.compute-1.amazonaws.com',
    password='vespucci 2015')
```

```python
ab = 'h3k4me2'
peptides = ['no_peptide', '10um_k99a', '100um_k99a', '0_1um_pcc', '10um_pcc']
```

```python
sql = '''select distinct on (p1.id) 
p1.id, chr."name" as chr_name, p1.start, p1."end",
p1.chromosome_id, 
p1.tag_count as tag_count_{peptide},
p1.id as id_{peptide},
p2.tag_count as tag_count_{other_1},
p2.id as id_{other_1},
p3.tag_count as tag_count_{other_2},
p3.id as id_{other_2},
p4.tag_count as tag_count_{other_3},
p4.id as id_{other_3},
p5.tag_count as tag_count_{other_4},
p5.id as id_{other_4},

ets.tag_count as ets_tag_count,
ets.id as ets_id,
batf.tag_count as batf_tag_count,
batf.id as batf_id,
junb.tag_count as junb_tag_count,
junb.id as junb_id,
cjun.tag_count as cjun_tag_count,
cjun.id as cjun_id,
junb.tag_count as junb_tag_count,
junb.id as junb_id,
irf4.tag_count as irf4_tag_count,
irf4.id as irf4_id

from chipseq.peak_cd4tcell_{ab}_{peptide}_24h_chip1 p1
join genome_reference_mm10.chromosome chr 
on p1.chromosome_id = chr.id

left outer join chipseq.peak_cd4tcell_{ab}_{other_1}_24h_chip1 p2
on p1.chromosome_id = p2.chromosome_id
and p1.start_end && p2.start_end
left outer join chipseq.peak_cd4tcell_{ab}_{other_2}_24h_chip1 p3
on p1.chromosome_id = p3.chromosome_id
and p1.start_end && p3.start_end
left outer join chipseq.peak_cd4tcell_{ab}_{other_3}_24h_chip1 p4
on p1.chromosome_id = p4.chromosome_id
and p1.start_end && p4.start_end
left outer join chipseq.peak_cd4tcell_{ab}_{other_4}_24h_chip1 p5
on p1.chromosome_id = p5.chromosome_id
and p1.start_end && p5.start_end

left outer join chipseq.peak_cd4tcell_ets1 ets
on p1.chromosome_id = ets.chromosome_id
and p1.start_end && ets.start_end
left outer join chipseq.peak_cd4tcell_th17_batf batf
on p1.chromosome_id = batf.chromosome_id
and p1.start_end && batf.start_end
left outer join chipseq.peak_cd4tcell_th17_junb junb
on p1.chromosome_id = junb.chromosome_id
and p1.start_end && junb.start_end
left outer join chipseq.peak_cd4tcell_th17_cjun cjun
on p1.chromosome_id = cjun.chromosome_id
and p1.start_end && cjun.start_end
left outer join chipseq.peak_cd4tcell_th17_jund jund
on p1.chromosome_id = jund.chromosome_id
and p1.start_end && jund.start_end
left outer join chipseq.peak_cd4tcell_th17_irf4 irf4
on p1.chromosome_id = irf4.chromosome_id
and p1.start_end && irf4.start_end

left outer join genome_reference_mm10.sequence_transcription_region reg
on p1.chromosome_id = reg.chromosome_id
and p1.start_end && reg.start_site_1000

where reg.id is {genic} NULL;
'''
```

```python
others_labels = ['other_1', 'other_2', 'other_3', 'other_4']
for i, current in enumerate(peptides):
    others = peptides[:i] + peptides[i+1:]
    others_d = OrderedDict(zip(others_labels, others))
    
    nongene_sql = sql.format(ab=ab, peptide=current, genic='', **others_d)
    gene_sql = sql.format(ab=ab, peptide=current, genic='NOT', **others_d)
    
    data = dataframe_from_query(nongene_sql, engine)
    data.to_csv('data/me2_and/enhancers_me2_and_{}.txt'.format(current), 
                sep='\t', header=True, index=False)
    data = dataframe_from_query(gene_sql, engine)
    data.to_csv('data/me2_and/promoters_me2_and_{}.txt'.format(current),
                sep='\t', header=True, index=False)
```

```python
# We want one unified set of all the peaks,
# but there is overlap between each of our sets.
# Take the record from the set with the biggest peak
# in each case where there are duplicates.
# This is not perfect, especially for the cases
# where one large peak covers several smaller,
# but it gives us a place to start.

for peak_type in ('enhancers','promoters'):
    keep_dfs = []
    for i, pep in enumerate(peptides):
        data = pd.io.parsers.read_csv('data/me2_and/{}_me2_and_{}.txt'.format(peak_type, pep), 
                                    sep='\t', header=0)
        data = data.fillna(0)
        print(len(data))
        key = 'tag_count_' + pep

        peptides_before = ['tag_count_' + o for o in peptides[:i]]
        peptides_after = ['tag_count_' + o for o in peptides[i+1:]]
        others = peptides_before + peptides_after
        data_mask = data[key] > data[others].max(axis=1)
        print(sum(data_mask))

        # If there is a tie, only take this row if it is the first instance of the tie.
        tie_mask = data[key] == data[others].max(axis=1)
        # Is this the first instance? Fill nulls for the case of the first column.
        first_mask = data[key] > data[peptides_before].max(axis=1).fillna(0)
        print(sum(tie_mask & first_mask))
        keep_dfs.append(data[data_mask | (tie_mask & first_mask)])

    keep = pd.concat(keep_dfs, join='outer', axis=0, ignore_index=True)
    print(len(keep))
    keep.head()
    keep.to_csv('data/me2_and/{}_me2_and_union.txt'.format(peak_type), 
                sep='\t', header=True, index=False)
```
