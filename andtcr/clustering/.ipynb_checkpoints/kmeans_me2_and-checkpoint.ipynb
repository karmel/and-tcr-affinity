```python
'''
Generate a basic clustering for motif finding, patterning, etc.

Use enhancers only here.
'''
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import scale
```

```python
sns.set_context('poster')
```

```python
data = pd.io.parsers.read_csv('../clustering/data/me2_and/enhancers_me2_and_union.txt', 
                                    sep='\t', 
                                    header=0)

data = data.fillna(0)
```

```python
peptides = ['no_peptide', '10um_k99a', '100um_k99a', '0_1um_pcc', '10um_pcc']
prefix = 'tag_count_'
fields = [prefix + p for p in peptides]
```

```python
# Extract fields of interest and normalize to unit norm row-wise
subdata = data[fields].div(data[fields].max(axis=1), axis=0)
subdata.head()
```

```python
krange = range(3,9)
for k in krange:
    kmodel = KMeans(k)
    kmodel.fit(subdata)
    
    subdata['label_{}'.format(k)] = kmodel.labels_
```

```python
subdata.head()
```

```python
for k in krange:
    clustered = subdata.sort(columns=['label_{}'.format(k),fields[0]])
    plt.figure()
    plt.pcolor(clustered[fields], cmap='coolwarm')
    plt.colorbar()
    plt.ylim([0,len(clustered)])
    title = 'Clustered Enhancers: {} clusters'.format(k)
    plt.title(title)
    plt.xticks([x + 0.5 for x in range(0, len(peptides))],
                peptides, rotation=90)
    plt.yticks([clustered['label_{}'.format(k)].tolist().index(x) for x in range(0,k)],
               ['Cluster {}'.format(x + 1) for x in range(0,k)])
    plt.savefig('figures/{}.png'.format(title.replace(' ','_')),
                bbox_inches='tight')
```

```python
# Output cluster files
for k in krange:
    clustered = subdata.sort(columns=['label_{}'.format(k),fields[0]])
    ordered_data = data.loc[clustered.index]
    ordered_data['label'] = clustered['label_{}'.format(k)]
    ordered_data.to_csv('data/clusters/kmeans_{}_me2_and.txt'.format(k), 
                        sep='\t', header=True, index=False)
```
