```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set_context('poster')
```

```python
peptides = ['no_peptide', '10um_k99a', '100um_k99a', '0_1um_pcc', '10um_pcc']
```

```python
for x in range(0, 4):
    data = []
    sections = ('bottom','top')
    for section in sections:
        fname = 'pca_pc{}_{}_me2_and_histogram'.format(x + 1, section)
        predata = pd.io.parsers.read_csv('data/me2_and/histograms/{}.txt'.format(fname), 
                        sep='\t', header=0, index_col=0)
        #predata = predata.apply(np.log2, axis=0)
        predata = predata.fillna(0)
        data.append(predata)

    both = pd.concat(data, join='outer', axis=0, ignore_index=True)
    ax = plt.subplot(111)
    plt.pcolor(both, cmap='Blues', vmax=8)
    
    plt.axhline(y=len(data[0]), color='black', linewidth=1)
    plt.xlim(0, both.shape[1])
    plt.ylim(0, both.shape[0])
    
    xticks = range(both.shape[1]//len(peptides)//2, both.shape[1], both.shape[1]//len(peptides))
    plt.xticks(xticks, peptides)
    
    yticks = range(both.shape[0]//len(sections)//2, both.shape[0], both.shape[0]//len(sections))
    plt.yticks(yticks, sections)
    
    title = 'H3K4me2 PC{}- Top and Bottom 10%'.format(x + 1)
    plt.title(title)
    plt.savefig('figures/{}.png'.format(title.replace(' ', '_')), bbox_inches='tight')
    
```

```python
for x in range(0,4):
    data = []
    for section in ('bottom','top'):
        fname = 'pca_pc{}_{}_me2_and_ets_ap1_histogram'.format(x + 1, section)
        predata = pd.io.parsers.read_csv('data/me2_and/histograms/{}.txt'.format(fname), 
                        sep='\t', header=0, index_col=0)
        predata = predata.fillna(0)
        data.append(predata)

    both = pd.concat(data, join='outer', axis=0, ignore_index=True)
    plt.figure()
    plt.pcolor(both, cmap='Blues', vmax=20)
    plt.colorbar()
    plt.xlim(0, both.shape[1])
    plt.ylim(0, both.shape[0])
    
```
