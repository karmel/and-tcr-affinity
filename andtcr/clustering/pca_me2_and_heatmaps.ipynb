```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set_context('poster')
```

```python
peptides = ['no_peptide', '10um_k99a', '100um_k99a', '0_1um_pcc', '10um_pcc']
```

```python
for x in range(0,4):
    data = []
    for section in ('bottom','top'):
        fname = 'pca_pc{}_{}_me2_and_histogram'.format(x + 1, section)
        predata = pd.io.parsers.read_csv('data/me2_and/histograms/{}.txt'.format(fname), 
                        sep='\t', header=0, index_col=0)
        #predata = predata.apply(np.log2, axis=0)
        predata = predata.fillna(0)
        data.append(predata)

    both = pd.concat(data, join='outer', axis=0, ignore_index=True)
    plt.figure()
    plt.pcolor(both, cmap='Blues', vmax=10)
    plt.colorbar()
    plt.xlim(0, both.shape[1])
    plt.ylim(0, both.shape[0])
    
```
