```python
'''
Use PCA to evaluate the axes of variation across our five conditions.
'''
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.decomposition import PCA
```

```python
sns.set_context('poster')
```

```python
data = pd.io.parsers.read_csv('../clustering/data/me2_and/enhancers_me2_and_union.txt', 
                                    sep='\t', 
                                    header=0)

data = data.fillna(0)
```

```python
peptides = ['no_peptide', '10um_k99a', '100um_k99a', '0_1um_pcc', '10um_pcc']
labels = ['No Peptide', '10uM K99A', '100uM K99A', '0.1uM PCC', '10uM PCC']
prefix = 'tag_count_'
fields = [prefix + p for p in peptides]
```

```python
# Extract fields of interest and normalize to unit norm row-wise
subdata = data[fields].div(data[fields].max(axis=1), axis=0)
subdata.head()
```

```python
# Filter out genes that aren't changing.
def check_changing(row):
    if row.max() - row.min() < row.min():
        return False
    return True
filtered = subdata[subdata.apply(check_changing, axis=1)]
print(len(filtered), len(subdata))
filtered.head()
```

```python
transpose = subdata.transpose()
pca = PCA(n_components=5)
X_r = pca.fit(transpose).transform(transpose)
print('Explained variance ratio (first five components): '
      + str(pca.explained_variance_ratio_))
X_r
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X_r[:, 0].tolist(), X_r[:, 1].tolist(), X_r[:, 2].tolist(), s=100)
plt.title('PCA: H3K4me2 by Condition')
plt.xlabel('PC1 (29% of variance)')
plt.ylabel('PC2 (25% of variance)')
ax.set_zlabel('PC3 (23% of variance)')

margin = 2
for i, txt in enumerate(labels):
    ax.text(X_r[i,0] + margin, 
            X_r[i,1] + margin, 
            X_r[i,2] + margin,
            txt, size=14)

plt.savefig('figures/PCA_H3K4me2_by_Condition_3D.png',
            bbox_inches='tight')
```

```python
fig = plt.figure()
plt.scatter(X_r[:, 0].tolist(), X_r[:, 1].tolist(), s=100)
plt.title('PCA: H3K4me2 by Condition')
plt.xlabel('PC1 (29% of variance)')
plt.ylabel('PC2 (25% of variance)')

margin = 2
for i, txt in enumerate(labels):
    plt.annotate(txt, (X_r[i,0] + margin, 
            X_r[i,1] + margin),
            size=14)

plt.savefig('figures/PCA_H3K4me2_by_Condition.png',
            bbox_inches='tight')

plt.axhline(color='black', linewidth=1)
plt.axvline(color='black', linewidth=1)
plt.savefig('figures/PCA_H3K4me2_by_Condition_axes.png',
            bbox_inches='tight')
```

```python
pd.DataFrame(pca.transform(transpose), index=fields)
```

```python
# We want to get the top and bottom features that make up each component.
components = pd.DataFrame(pca.components_)
for x in range(0,4):
    pc = components.loc[x]
    pc.sort()
    num_elem = len(pc)
    bottom = pc.index[:num_elem//10]
    top = pc.index[9*num_elem//10:]
    
    data.loc[bottom].to_csv('data/clusters/pca_pc{}_bottom_me2_and.txt'.format(x + 1), 
                        sep='\t', header=True, index=False)
    data.loc[top].to_csv('data/clusters/pca_pc{}_top_me2_and.txt'.format(x + 1), 
                        sep='\t', header=True, index=False)
```

```python
# Plot percent variance explained by the first four PCs
num = 4
span = range(0,num)
ratios = pca.explained_variance_ratio_[:num]*100
plt.figure()
plt.bar([x + .2 for x in span], ratios)
plt.xlim([0, num + .2])
plt.ylabel('Percent of variance explained')
plt.xticks([x + .6 for x in span], ['PC{}'.format(x + 1) for x in span])
title = 'Variance Percentages by Principal Component'
plt.title(title)
plt.savefig('figures/{}.png'.format(title.replace(' ', '_')), bbox_inches='tight')
```
