```python
'''
For any given promoter me2 peak that is changing, 
find the nearest neighbor non-genic me2 peaks.

Use chr and transcription location as weighted dimensions, too.
'''
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.neighbors import NearestNeighbors
from sklearn.preprocessing import scale
```

```python
sns.set_context('poster')
```

```python
genes = pd.io.parsers.read_csv('data/me2_and/promoters_me2_and_union.txt', 
                                    sep='\t', 
                                    header=0)
nongenes = pd.io.parsers.read_csv('data/me2_and/enhancers_me2_and_union.txt', 
                                    sep='\t', 
                                    header=0)
```

```python
peptides = ['no_peptide', '10um_k99a', '100um_k99a', '0_1um_pcc', '10um_pcc']
fields = ['tag_count_' + p for p in peptides]
```

```python
# We only care about changing genes.
fold = 2
genes_mask = genes[fields].min(axis=1)*fold <= genes[fields].max(axis=1)
genes_chg = genes[genes_mask]
print(len(genes_chg))
genes_chg.head()
```

```python
# Extract columns of interest and scale row-wise to unit norm.
genes_v = genes_chg[fields]
nongenes_v = nongenes[fields]

for data in (genes_v, nongenes_v):
    max_rowwise = data.max(axis=1)
    for col in data.columns:
        data[col] = data[col]/max_rowwise
    
genes_v.head()
```

```python
# Add back in chromosome column, and scale so that it has greater weight.
chr_weight = 100
genes_v['chr'] = genes_chg['chromosome_id']/genes_chg['chromosome_id'].max() * chr_weight
nongenes_v['chr'] = nongenes['chromosome_id']/nongenes['chromosome_id'].max() * chr_weight

# Add back in transcription start and end, and scale across all samples
bp_weight = 100
genes_v['start'] = genes_chg['start']/genes_chg['start'].max()*bp_weight
genes_v['end'] = genes_chg['end']/genes_chg['end'].max()*bp_weight
nongenes_v['start'] = nongenes['start']/nongenes['start'].max()*bp_weight
nongenes_v['end'] = nongenes['end']/nongenes['end'].max()*bp_weight
```

```python
# We want to set a radius rather than number of neighbors,
# as not all genes will have the same number of "enhancers"
radii = np.arange(0,1,.05)
              

nbrs = NearestNeighbors(algorithm='kd_tree').fit(nongenes_v)
one_nbr = []
for i, r in enumerate(radii):
    distances, indices = nbrs.radius_neighbors(genes_v, radius=r)
    
    # Draw histogram of number of neighbors found.
    counts = [len(ind_set) for ind_set in indices]
    cols = 4
    ax = plt.subplot(np.ceil(len(radii)/cols), cols, i + 1)
    plt.hist(counts)
    plt.title('Radius = {}'.format(r), fontsize=10)
    ax.tick_params(axis='x', labelsize=10)
    ax.tick_params(axis='y', labelsize=10)

    # How many genes have exactly one neighbor?
    one_nbr.append(len([x for x in counts if x == 1]))
plt.tight_layout()
    
plt.figure()
plt.plot(radii, one_nbr, 'o-')
```

```python
# HiC data is our True label. Evaluate rate of retrieving True interactions.
radii = np.arange(0,.5,.025)
              
matrix = [] # @todo
resolution = 10000

nbrs = NearestNeighbors(algorithm='kd_tree').fit(nongenes_v)
mean_rates = []
for i, r in enumerate(radii):
    distances, indices = nbrs.radius_neighbors(genes_v, radius=r)
    
    ixn_rates = []
    for (idx, gene), ind_set in zip(genes_v.iterrows(), indices):
        # Find the gene in the interaction matrix
        start = genes_chg.loc[idx]['transcription_start']//resolution * resolution
        chrom = genes_chg.loc[idx]['chr_name']
        gene_column = matrix[chrom + '-' + start]
        
        # For each predicted partner, what is our interaction expectation?
        ng_indices = nongenes_v.index[ind_set]
        for ind in ng_indices:
            start = nongenes.loc[ind]['transcription_start']//resolution * resolution
            chrom = nongenes.loc[ind]['chr_name']
            ixn_rates.append(gene_column.loc[chrom + '-' + start].tolist()[0])
            
    # Draw histogram of interaction rates from HiC for predicted partners.
    cols = 4
    ax = plt.subplot(np.ceil(len(radii)/cols), cols, i + 1)
    plt.hist(ixn_rates)
    plt.title('Radius = {}'.format(r), fontsize=10)
    ax.tick_params(axis='x', labelsize=10)
    ax.tick_params(axis='y', labelsize=10)

    # What is the mean
    mean_rates.append(np.mean(ixn_rates))
plt.tight_layout()
    
plt.figure()
plt.plot(radii, mean_rates, 'o-')
```
