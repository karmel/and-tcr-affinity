```python
'''
Do top versus bottom genes show preferential "spreading" of dimethyl
along the ggene body after activation?
'''
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set_context('poster')
```

```python
# First, we want to generate a file that has gene promoter locations,
# and another that has bodies less the promoter.
genes = pd.io.parsers.read_csv('../rna/data/ap1_per_gene.txt', 
                                    sep='\t', 
                                    header=0, index_col=0)
genes.head()
```

```python
# Some math to get the starts of sense genes and the ends - 1kb of the antisense
size = 1000
genes['promoter_start'] = (genes['transcription_end'] - size)*genes['strand'] \
                            - genes['transcription_start']*(genes['strand'] - 1)
genes['promoter_start'].head()
```

```python
genes['promoter_end'] = genes['transcription_end']*genes['strand'] \
                            - (genes['transcription_start'] + size)*(genes['strand'] - 1)
genes['promoter_end'].head()
```

```python
# Now, bodies
genes['body_start'] = genes['transcription_start']*genes['strand'] \
                            - (genes['promoter_end'] + 1)*(genes['strand'] - 1)
genes['body_start'].head()
```

```python
genes['body_end'] = (genes['promoter_start'] - 1)*genes['strand'] \
                            - genes['transcription_end']*(genes['strand'] - 1)
genes['body_end'].head()
```

```python
# A few checks
print(sum(genes['promoter_end'] - genes['promoter_start'] != size))
print(sum(genes['body_end'] - genes['body_start'] + size + 1 != genes['transcription_end'] - genes['transcription_start']))
```

```python
# Output files.
genes[['chr_name', 'promoter_start', 'promoter_end', 'strand']].to_csv(
    'output/gene_sections/gene_promoters.txt', sep='\t',
    header=['chr_name', 'start', 'end', 'strand'], 
    index=True)
genes[['chr_name', 'body_start', 'body_end', 'strand']].to_csv(
    'output/gene_sections/gene_bodies.txt', sep='\t',
    header=['chr_name', 'start', 'end', 'strand'], 
    index=True)
```

```python
'''
Those files are used to count tags with Homer:

for f in *
do
basename=`echo $f | sed 's/.txt//'`
annotatePeaks.pl $f mm10 -size given -d ../tag-dirs/AND-CD4TCell-ChIP-H3K4me2-* > ${basename}_tag_counts.txt
done
'''
```

```python
sections = []
section_names = ('promoters', 'bodies')
for section in section_names:
    data = pd.io.parsers.read_csv('data/gene_{}_tag_counts.txt'.format(section), 
                                        sep='\t', 
                                        header=0, index_col=0)

    data = data.fillna(0) 

    # Rename our columns of interest
    labels = ['No Peptide', '10uM K99A', '100uM K99A', '0.1uM PCC', '10uM PCC']
    last_five_columns = [labels[3], labels[2], labels[1], labels[4], labels[0]]

    data.columns = data.columns[:-5].tolist() + last_five_columns
    sections.append(data)
```

```python
# Get genes from PC1
top_genes = pd.io.parsers.read_csv('../rna/unsupervised/output/clusters/pca_pc1_top_rna_and.txt', 
                                    sep='\t', 
                                    header=0, index_col=0)
bottom_genes = pd.io.parsers.read_csv('../rna/unsupervised/output/clusters/pca_pc1_bottom_rna_and.txt', 
                                    sep='\t', 
                                    header=0, index_col=0)
top_genes.head()
```

```python
tops, bottoms = [], []
for section in sections:
    tops.append(section[section.index.isin(top_genes.index)].loc[top_genes.index])
    bottoms.append(section[section.index.isin(bottom_genes.index)].loc[bottom_genes.index])
    
```

```python
[len(t) for t in tops]
```

```python
# Get ratio of body tags to promoter tags for each condition.
top_ratios = tops[1][labels].div(tops[0][labels].replace(0,1), axis=0)
bottom_ratios = bottoms[1][labels].div(bottoms[0][labels].replace(0,1), axis=0)
bottom_ratios.head()
```

```python
# Set up colors
colorset = sns.color_palette("husl", 5)[::-1]
```

```python
# Box plots of conditions
bp = plt.boxplot([top_ratios[label] for label in labels], patch_artist=True)

# Colors.
for i in range(0, len(labels)):
    outline_col = '#666666'
    plt.setp(bp['boxes'][i], edgecolor=outline_col)
    plt.setp(bp['boxes'][i], linewidth=3)
    plt.setp(bp['boxes'][i], facecolor=colorset[i])
    plt.setp(bp['medians'][i], color=outline_col)
    plt.setp(bp['whiskers'][i*2], color=outline_col)
    plt.setp(bp['whiskers'][i*2+1], color=outline_col)
    plt.setp(bp['caps'][i*2], color=outline_col)
    plt.setp(bp['caps'][i*2+1], color=outline_col)

    

plt.ylim([0,30])

plt.ylabel('Ratio of gene body tags to promoter tags')
plt.xticks(range(1,len(labels)+1), labels)

title = 'Gene body-to-promoter H3K4me2 ratios for activation signature genes'
plt.title(title)

plt.savefig('figures/{}.png'.format(title.replace(' ','_')),
            bbox_inches='tight')
```

```python
# Box plots of conditions
bp = plt.boxplot([bottom_ratios[label] for label in labels], patch_artist=True)

# Colors.
for i in range(0, len(labels)):
    outline_col = '#666666'
    plt.setp(bp['boxes'][i], edgecolor=outline_col)
    plt.setp(bp['boxes'][i], linewidth=3)
    plt.setp(bp['boxes'][i], facecolor=colorset[i])
    plt.setp(bp['medians'][i], color=outline_col)
    plt.setp(bp['whiskers'][i*2], color=outline_col)
    plt.setp(bp['whiskers'][i*2+1], color=outline_col)
    plt.setp(bp['caps'][i*2], color=outline_col)
    plt.setp(bp['caps'][i*2+1], color=outline_col)

    

plt.ylim([0,40])

plt.ylabel('Ratio of gene body tags to promoter tags')
plt.xticks(range(1,len(labels)+1), labels)

title = 'Gene body-to-promoter H3K4me2 ratios for bottom 10% genes'
plt.title(title)

plt.savefig('figures/{}.png'.format(title.replace(' ','_')),
            bbox_inches='tight')
```

```python
# Compare promoter and body counts to No Pep directly
top_promoter = tops[0][labels[1:]].div(tops[0][labels[0]], axis=0)
top_body = tops[1][labels[1:]].div(tops[1][labels[0]], axis=0)
top_promoter.head()
```

```python
# Box plots of conditions
bp = plt.boxplot([top_body[label] for label in labels[1:]], patch_artist=True)

# Colors.
for i in range(0, len(labels[1:])):
    outline_col = '#666666'
    plt.setp(bp['boxes'][i], edgecolor=outline_col)
    plt.setp(bp['boxes'][i], linewidth=3)
    plt.setp(bp['boxes'][i], facecolor=colorset[1:][i])
    plt.setp(bp['medians'][i], color=outline_col)
    plt.setp(bp['whiskers'][i*2], color=outline_col)
    plt.setp(bp['whiskers'][i*2+1], color=outline_col)
    plt.setp(bp['caps'][i*2], color=outline_col)
    plt.setp(bp['caps'][i*2+1], color=outline_col)

    

plt.ylim([0,5])

plt.ylabel('Ratio of gene body tags to No Peptide gene body tags')
plt.xticks(range(1,len(labels[1:])+1), labels[1:])

title = 'Activated to No Peptide H3K4me2 ratios in gene bodies of activation signature genes'
plt.title(title)

plt.savefig('figures/{}.png'.format(title.replace(' ','_')),
            bbox_inches='tight')
```

```python
# Output top and bottom genes sorted by body-to-promoter ratio
# We want to use this for a heatmap, so only take the first _size_ basepairs,
# starting from the promoter
size = 5000

top_sorted = top_ratios.sort(labels[4], ascending=False)
bottom_sorted = bottom_ratios.sort(labels[4], ascending=False)

genes['heatmap_start'] = (genes['transcription_end'] - size)*genes['strand'] \
                            - genes['transcription_start']*(genes['strand'] - 1)
genes['heatmap_end'] = genes['transcription_end']*genes['strand'] \
                            - (genes['transcription_start'] + size)*(genes['strand'] - 1)

heatmap_locs = genes[['chr_name', 'heatmap_start', 'heatmap_end', 'strand']]

heatmap_locs.loc[top_sorted.index].to_csv('output/gene_sections/body_promoter_ratio_top_genes.txt',
                                       sep='\t', header=True, index=True)
heatmap_locs.loc[bottom_sorted.index].to_csv('output/gene_sections/body_promoter_ratio_bottom_genes.txt',
                                       sep='\t', header=True, index=True)
bottom_sorted.head()
```

```python
# Now get histograms to generate heatmaps
top_hist = pd.io.parsers.read_csv('data/body_promoter_ratio_top_genes_histogram.txt', 
                                        sep='\t', 
                                        header=0, index_col=0)
bottom_hist = pd.io.parsers.read_csv('data/body_promoter_ratio_bottom_genes_histogram.txt', 
                                        sep='\t', 
                                        header=0, index_col=0)
```

```python
# Flip the histograms for antisense genes so that everything lines up.
# Note that we have to flip each separate condition within itself.
# Note further that re-assigning values by row, slice does not seem to work
# with pandas, so we go cell-by-cell.
for hist in (top_hist, bottom_hist):
    for i in hist.index:
        if genes.loc[i].strand == 1:
            per_cond = len(hist.columns)//len(labels)
            for j, col in enumerate(hist.columns):
                curr_cond = j//per_cond
                #print(curr_cond)
                new_j = curr_cond*per_cond + per_cond - (j % per_cond) - 1
                hist.loc[i, col] = hist.loc[i, hist.columns[new_j]]
                #print(hist.columns[new_j], )

            #print(hist.loc[i])
top_hist.head()
```

```python
data = top_hist.loc[top_hist.index[::-1]]
ax = plt.subplot(111)
plt.pcolor(data, cmap='Blues', vmax=.25)

plt.xlim(0, data.shape[1])
plt.ylim(0, data.shape[0])

xticks = range(data.shape[1]//len(labels)//2, data.shape[1], data.shape[1]//len(labels))
plt.xticks(xticks, labels)

title = 'H3K4me2 at genes in top 10% of PC1'
plt.title(title)
plt.savefig('figures/{}.png'.format(title.replace(' ', '_')), bbox_inches='tight')
```

```python
data = bottom_hist.loc[bottom_hist.index[::-1]]
ax = plt.subplot(111)
plt.pcolor(data, cmap='Blues', vmax=.25)

plt.xlim(0, data.shape[1])
plt.ylim(0, data.shape[0])

xticks = range(data.shape[1]//len(labels)//2, data.shape[1], data.shape[1]//len(labels))
plt.xticks(xticks, labels)

title = 'H3K4me2 at genes in bottom 10% of PC1'
plt.title(title)
plt.savefig('figures/{}.png'.format(title.replace(' ', '_')), bbox_inches='tight')
```

```python
# Set up colors
colorset = sns.color_palette("husl", len(labels))[::-1]
```

```python
# Instead of a heatmap, try cumulative plots.
data = top_hist.mean(axis=0)
ax = plt.subplot(111)

for i, label in enumerate(labels):
    plt.plot(data[i*per_cond:(i+1)*per_cond], label=label, color=colorset[i])

plt.legend()
plt.xlabel('Base-pairs from TSS')
plt.ylabel('Mean H3K4me2 tags')

plt.xlim([0,20])
plt.xticks(range(0,20,2), range(0,4000,400))

title = 'H3K4me2 spread at genes in top 10% of PC1'
plt.title(title)
plt.savefig('figures/{}.png'.format(title.replace(' ', '_')), bbox_inches='tight')
```
