```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import scipy
```

```python
sns.set_context('poster')
```

```python
ab = 'h3k27ac'
data = pd.io.parsers.read_csv('../offstate/data/union_peaks_1um_{}.txt'.format(ab), 
                sep='\t', header=0)
data = data.fillna(0)

# Rename our columns of interest
labels = ['No Peptide', '1uM K99A', '1uM PCC']
last_columns = [labels[1], labels[2], labels[0]]

data.columns = data.columns[:-len(last_columns)].tolist() + last_columns

data.head()
```

```python
# Enhancers only
data = data[data['Annotation'].str.contains('promoter') == False]
#data = data[data['Distance to TSS'] < 5000]
data.describe()
```

```python
min_tags = 10
```

```python
# Activated only?
fold = 2
act = labels[1:]
act_only = data[(data[act].max(axis=1) > data[labels[0]]*fold) &
              (data[act].max(axis=1) > min_tags)]
print('Activated but not {} enhancers'.format(labels[0]), len(act_only))
act_only[labels].describe()
```

```python
# Groups of enhancers of interest:
# de novo, next to act sig, next to bottom 10%
act_only.to_csv('output/subgroups/high_affinity_{}x_enhancers_{}.txt'.format(fold, ab), 
              sep='\t', header=True, index=False)
```

```python
# Get genes from PC1
top_genes = pd.io.parsers.read_csv('../rna/unsupervised/output/clusters/pca_pc1_top_rna_and.txt', 
                                    sep='\t', 
                                    header=0, index_col=0)
bottom_genes = pd.io.parsers.read_csv('../rna/unsupervised/output/clusters/pca_pc1_bottom_rna_and.txt', 
                                    sep='\t', 
                                    header=0, index_col=0)
top_genes.head()
```

```python
top_enh = data[data['Gene Name'].isin(top_genes['gene_name'])]
top_enh.to_csv('output/subgroups/enhancers_near_top_pc1_genes_{}.txt'.format(ab), 
              sep='\t', header=True, index=False)
top_enh.describe()
```

```python
bottom_enh = data[data['Gene Name'].isin(bottom_genes['gene_name'])]
bottom_enh.to_csv('output/subgroups/enhancers_near_bottom_pc1_genes_{}.txt'.format(ab), 
              sep='\t', header=True, index=False)
bottom_enh.describe()
```

```python
# Set up colors
colorset = sns.color_palette("husl", len(labels))[::-1]
```

```python
# Box plots of conditions
bp = plt.boxplot([top_enh[label] for label in labels], patch_artist=True)

# Colors.
for i in range(0, len(labels)):
    outline_col = '#666666'
    plt.setp(bp['boxes'][i], edgecolor=outline_col)
    plt.setp(bp['boxes'][i], linewidth=3)
    plt.setp(bp['boxes'][i], facecolor=colorset[i])
    plt.setp(bp['medians'][i], color=outline_col)
    plt.setp(bp['whiskers'][i*2], color=outline_col)
    plt.setp(bp['whiskers'][i*2+1], color=outline_col)
    plt.setp(bp['caps'][i*2], color=outline_col)
    plt.setp(bp['caps'][i*2+1], color=outline_col)

    

plt.ylim([0,150])

plt.ylabel('Tag count')
plt.xticks(range(1,len(labels)+1), labels)

title = 'H3K27Ac tag counts at enhancers near top 10% of PC1'
plt.title(title)

plt.savefig('figures/{}.png'.format(title.replace(' ','_')),
            bbox_inches='tight')
```

```python
# Do denovo show enrichment for top genes?
top_ct = sum(act_only['Gene Name'].isin(top_genes['gene_name']))
bottom_ct = sum(act_only['Gene Name'].isin(bottom_genes['gene_name']))
top_ct, bottom_ct
```

```python

act_only_not_top = len(act_only) - top_ct
existing_top = len(top_genes) - top_ct
existing_not_top = len(data['Gene Name'].unique())
scipy.stats.chi2_contingency([[top_ct, existing_top], 
                             [act_only_not_top, existing_not_top]])
```
