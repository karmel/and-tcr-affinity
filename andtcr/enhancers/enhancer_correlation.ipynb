```python
'''
How do no-pep enhancers compare with the other two conditions?
'''
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
```

```python
sns.set_context('poster')
```

```python
data = pd.io.parsers.read_csv('data/union_peaks_five_conditions.txt', 
                                    sep='\t', 
                                    header=0)

data = data.fillna(0)

# Rename our columns of interest
labels = ['No Peptide', '10uM K99A', '100uM K99A', '0.1uM PCC', '10uM PCC']
last_five_columns = [labels[3], labels[2], labels[1], labels[4], labels[0]]

data.columns = data.columns[:-5].tolist() + last_five_columns
```

```python
limit = 2000
```

```python
for y in range(1, len(labels)):
    x = 0
    r = pearsonr(data[labels[x]], data[labels[y]])

    plt.figure()
    sns.regplot(labels[x],labels[y], data,
                label='Linear fit (r={:.2f})'.format(r[0]))
    plt.xlim([0,limit])
    plt.ylim([0,limit])

    plt.legend()

    plt.xlabel('{} Enhancer Tag Count'.format(labels[x]))
    plt.ylabel('{} Enhancer Tag Count'.format(labels[y]))

    title = '{} versus {} enhancer correlation'.format(labels[x], labels[y])
    plt.title(title)

    plt.savefig('figures/{}.png'.format(title.replace(' ','_')), 
                bbox_inches='tight')
```
