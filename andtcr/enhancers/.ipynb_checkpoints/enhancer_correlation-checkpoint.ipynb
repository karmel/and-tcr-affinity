```python
'''
How do no-pep enhancers compare with the other two conditions?
'''
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
```

```python
sns.set_context('poster')
```

```python
data = pd.io.parsers.read_csv('data/union_peaks_five_conditions.txt', 
                                    sep='\t', 
                                    header=0)

data = data.fillna(0)

# Rename our columns of interest
labels = ['No Peptide', '10uM K99A', '100uM K99A', '0.1uM PCC', '10uM PCC']
last_five_columns = [labels[3], labels[2], labels[1], labels[4], labels[0]]

data.columns = data.columns[:-5].tolist() + last_five_columns
```

```python
colorset = sns.color_palette("husl", 5)[::-1]
```

```python
fold = 2
min_tags = 10
for y in range(1, len(labels)):
    x = 0
    r = pearsonr(data[labels[x]], data[labels[y]])

    # We want to separately color increasing with activation.
    data = data[data[[labels[x], labels[y]]].max(axis=1) >= min_tags]
    act_only = data[data[labels[y]] > data[labels[x]]*fold]
    rest = data[data[labels[y]] <= data[labels[x]]*fold]
    
    fig = plt.figure()
    ax = fig.add_subplot(111)
    plt.plot(rest[labels[x]], rest[labels[y]], '.',
             label='Poised enhancers (r={:.2f})'.format(r[0]))
    plt.plot(act_only[labels[x]], act_only[labels[y]], '.', color=colorset[y],
             label='> {}x tags in {}'.format(fold, labels[y]))

    plt.xlim([1,plt.xlim()[1]])
    plt.ylim([1,plt.ylim()[1]])

    ax.set_xscale('log')
    ax.set_yscale('log')
    
    plt.legend(loc='upper left')

    plt.xlabel('{} Enhancer Tag Count'.format(labels[x]))
    plt.ylabel('{} Enhancer Tag Count'.format(labels[y]))

    title = '{} versus {} enhancer correlation'.format(labels[x], labels[y])
    plt.title(title)

    plt.savefig('figures/{}.png'.format(title.replace(' ','_')), 
                bbox_inches='tight')
```
