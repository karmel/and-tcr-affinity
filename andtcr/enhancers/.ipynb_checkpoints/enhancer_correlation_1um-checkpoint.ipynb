```python
'''
How do no-pep enhancers compare with the other two conditions?
'''
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
```

```python
ab = 'h3k4me2'
data = pd.io.parsers.read_csv('data/union_peaks_1um_{}.txt'.format(ab),
                sep='\t', header=0)
data = data.fillna(0)

# Rename our columns of interest
labels = ['No Peptide', '1uM K99A', '1uM PCC']
last_columns = [labels[1], labels[2], labels[0]]

data.columns = data.columns[:-len(last_columns)].tolist() + last_columns

data.head()
```

```python
limit = 2000
```

```python
r = pearsonr(data[labels[0]], data[labels[1]])

plt.plot(data[labels[0]],data[labels[1]], '.',
            label='r={:.2f}'.format(r[0]))
plt.xlim([0,limit])
plt.ylim([0,limit])

plt.legend()

plt.xlabel('No Peptide Enhancer Tag Count')
plt.ylabel('1uM K99A Enhancer Tag Count')

title = 'No Peptide versus 1uM K99A enhancer correlation'
plt.title(title, y=1.01)

plt.savefig('figures/{}'.format(title.replace(' ','_')), 
            bbox_inches='tight')
```

```python
r = pearsonr(data[labels[0]], data[labels[2]])

plt.plot(data[labels[0]],data[labels[2]], '.',
            label='r={:.2f}'.format(r[0]))

plt.xlim([0,limit])
plt.ylim([0,limit])

plt.legend()

plt.xlabel('No Peptide Enhancer Tag Count')
plt.ylabel('1uM PCC Enhancer Tag Count')

title = 'No Peptide versus 1uM PCC enhancer correlation'
plt.title(title, y=1.01)

plt.savefig('figures/{}'.format(title.replace(' ','_')), 
            bbox_inches='tight')
```

```python
r = pearsonr(data[labels[1]], data[labels[2]])

plt.plot(data[labels[1]],data[labels[2]], '.',
            label='r={:.2f}'.format(r[0]))

plt.xlim([0,limit])
plt.ylim([0,limit])

plt.legend()

plt.xlabel('1uM K99A Enhancer Tag Count')
plt.ylabel('1uM PCC Enhancer Tag Count')

title = '1uM K99A versus 1uM PCC enhancer correlation'
plt.title(title, y=1.01)

plt.savefig('figures/{}'.format(title.replace(' ','_')), 
            bbox_inches='tight')
```
