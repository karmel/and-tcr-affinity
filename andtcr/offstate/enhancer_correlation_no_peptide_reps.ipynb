```python
'''
How do no-pep enhancer runs compare to each other?
'''
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
```

```python
sns.set_context('poster')
```

```python
data = pd.io.parsers.read_csv('data/enhancers_me2_and_union_no_peptide.txt', 
                sep='\t', header=0)
data = data.fillna(0)
data.head()
```

```python
limit = 1000
```

```python
r = pearsonr(data['tag_count_no_peptide'], data['tag_count_1um_k99a'])

sns.regplot('tag_count_no_peptide','tag_count_1um_k99a', data,
            label='Linear fit (r={:.2f})'.format(r[0]))
plt.xlim([0,limit])
plt.ylim([0,limit])

plt.legend()

plt.xlabel('No Peptide Enhancer Tag Count')
plt.ylabel('1uM K99A Enhancer Tag Count')

title = 'No Peptide versus 1uM K99A enhancer correlation'
plt.title(title)

plt.savefig('figures/{}.png'.format(title.replace(' ','_')), 
            bbox_inches='tight')
```

```python
r = pearsonr(data['tag_count_no_peptide'], data['tag_count_1um_pcc'])
sns.regplot('tag_count_no_peptide','tag_count_1um_pcc', data,
            label='Linear fit (r={:.2f})'.format(r[0]))
plt.xlim([0,limit])
plt.ylim([0,limit])

plt.legend()

plt.xlabel('No Peptide Enhancer Tag Count')
plt.ylabel('1uM PCC Enhancer Tag Count')

title = 'No Peptide versus 1uM PCC enhancer correlation'
plt.title(title)

plt.savefig('figures/{}.png'.format(title.replace(' ','_')), 
            bbox_inches='tight')
```

```python
r = pearsonr(data['tag_count_1um_k99a'], data['tag_count_1um_pcc'])
sns.regplot('tag_count_1um_k99a','tag_count_1um_pcc', data,
            label='Linear fit (r={:.2f})'.format(r[0]))
plt.xlim([0,limit])
plt.ylim([0,limit])

plt.legend()

plt.xlabel('1uM K99A Enhancer Tag Count')
plt.ylabel('1uM PCC Enhancer Tag Count')

title = '1uM K99A versus 1uM PCC enhancer correlation'
plt.title(title)

plt.savefig('figures/{}.png'.format(title.replace(' ','_')), 
            bbox_inches='tight')
```
