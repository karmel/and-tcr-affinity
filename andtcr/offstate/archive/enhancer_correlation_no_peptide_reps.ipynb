```python
'''
How do no-pep enhancer runs compare to each other?
'''
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
```

```python
sns.set_context('poster')
```

```python
data = pd.io.parsers.read_csv('data/enhancers_me2_and_union_no_peptide.txt', 
                sep='\t', header=0)
data = data.fillna(0)
data = data[data.filter(like='tag_count').max(axis=1) > 20]
data.head()
```

```python
limit = 1000
dates = ['13_11_20', '13_11_25', '14_06_11']
```

```python
r = pearsonr(data['tag_count_' + dates[0]], data['tag_count_' + dates[1]])

sns.regplot('tag_count_' + dates[0],'tag_count_' + dates[1], data,
            label='Linear fit (r={:.2f})'.format(r[0]))
plt.xlim([0,limit])
plt.ylim([0,limit])

plt.legend()

plt.xlabel('No Peptide Enhancer Tag Count- Replicate 1')
plt.ylabel('No Peptide Enhancer Tag Count- Replicate 2')

title = 'No Peptide enhancer correlation- Replicates 1 and 2'
plt.title(title)

plt.savefig('figures/{}.png'.format(title.replace(' ','_')), 
            bbox_inches='tight')
```

```python
r = pearsonr(data['tag_count_' + dates[0]], data['tag_count_' + dates[2]])

sns.regplot('tag_count_' + dates[0],'tag_count_' + dates[2], data,
            label='Linear fit (r={:.2f})'.format(r[0]))
plt.xlim([0,limit])
plt.ylim([0,limit])

plt.legend()

plt.xlabel('No Peptide Enhancer Tag Count- Replicate 1')
plt.ylabel('No Peptide Enhancer Tag Count- Replicate 3')

title = 'No Peptide enhancer correlation- Replicates 1 and 3'
plt.title(title)

plt.savefig('figures/{}.png'.format(title.replace(' ','_')), 
            bbox_inches='tight')
```

```python
r = pearsonr(data['tag_count_' + dates[1]], data['tag_count_' + dates[2]])

sns.regplot('tag_count_' + dates[1],'tag_count_' + dates[2], data,
            label='Linear fit (r={:.2f})'.format(r[0]))
plt.xlim([0,limit])
plt.ylim([0,limit])

plt.legend()

plt.xlabel('No Peptide Enhancer Tag Count- Replicate 2')
plt.ylabel('No Peptide Enhancer Tag Count- Replicate 3')

title = 'No Peptide enhancer correlation- Replicates 2 and 3'
plt.title(title)

plt.savefig('figures/{}.png'.format(title.replace(' ','_')), 
            bbox_inches='tight')
```
